```python
### [코드 12-6]

def get_binary_number(decimal_number):
    assert isinstance(decimal_number, int)
    return bin(decimal_number)
print(get_binary_number(10))
print(get_binary_number("10"))
```

```python
### 연습문제 6 - 02

class Company:
    def __init__(self):
        self.work = True
        self.name = 'Jane'
        self.gender = 'woman'

    def retire(self):
        self.work = False
    
class Employee(Company):
    def __init__(self, name, gender):
        super().__init__()
        self.name = name
        self.gender = gender

    def introduce(self):
        if self.work == True:
            print('I got a job at a company')
            print(f'My name is {self.name}')
            print(f'I am a {self.gender}')
        if self.work == False:
            print('I left the company')
    
em = Employee('james', 'man')
em.retire()
em.introduce()
Employee('james', 'man').introduce()
```

```python
### 연습문제 6 - 08

class TV():
    def __init__(self, size, year, company):
        self.size = size
        self.year = year
        self.company = company
    def describe(self):
        print(self.company + "에서 만든" + self.year + "년형" + self.size + "인치" + "TV")

class Laptop(TV ):
    def describe(self):
        print(self.company + "에서 만든" + self.year + "년형" + self.size + "인치" + "TV")
    
LG_TV = TV("32", "2022", "LG")
LG_TV.describe()

samsung_microwave = Laptop("15", "2023", "samsung")
samsung_microwave.describe()
```

```python
### [코드 12-16]

# 필요한 라이브러리/모듈 임폴트
import os, random, datetime

os.makedirs("log", exist_ok=True)

if not os.path.exists("log/count_log.txt"):
    f = open("log/count_log.txt", 'w', encoding='utf8')
    f.write("기록이 시작됩니다.\n")
    f.close()

with open("log/count_log.txt", 'a', encoding='utf8') as f:
    for i in range(1, 11):
        stamp = str(datetime.datetime.now())
        value = random.random() * 1000000
        log_line = stamp + "\t" + str(value) + "값이 생성되었습니다." + "\n"
        f.write(log_line)
```
