```python
### loc 인덱서와 슬라이싱 사용

'''
# '번호' 컬럼부터 '등록일자' 컬럼까지 추출 --> 새로운 DataFrame 생성
'''

ns_book = ns_df.loc[:, '번호':'등록일자']

# 결과 확인하기
ns_book
```

## 공공 데이터를 이용하여 중복된 행 찾기

```python
### duplicated() 이용 --> 중복된 행 찾기

# duplicated() 함수 적용 --> boolean 배열 생성
duplicated_rows = ns_book.duplicated()
print(duplicated_rows)

print('-'*80)

# 중복된 행의 갯수 계산하기
print(f'중복된 행의 수 = {duplicated_rows.sum()}')
```

```python
### '도서명', '저자', 'ISBN' 컬럼 기준으로 중복된 행 찾기(1)

# duplicated(subset=[]) 함수 적용 --> boolean 배열 생성
duplicated_rows = ns_book.duplicated(subset=['도서명', '저자', 'ISBN'])

# 중복된 행의 수 계산기
print(f'중복된 행의 수 = {duplicated_rows.sum()}')
```

```python
### '도서명', '저자', 'ISBN' 컬럼 기준으로 중복된 행 찾기(2)

# duplicated(subset=[], keep=False) 함수 적용 --> boolean 배열 생성
dup_rows = ns_book.duplicated(subset=['도서명', '저자', 'ISBN'], keep=False)

# 중복된 행의 수 계산기
print(f'중복된 행의 수 = {dup_rows.sum()}')
```

```python
### loc 인덱서와 boolean 배열 이용 --> 중복된 행 추출

# boolean 배열 확인
print(dup_rows)

print('-'*80)

# ns_book.loc[dup_rows, :]
ns_book3 = ns_book.loc[dup_rows, :]

# 결과 확인하기
ns_book3
```

```python
### 중복된 행 제거

# drop_duplicates(subset, keep, inplace, ignore_index) 사용
del_duplicates_book = ns_book.drop_duplicates(subset=['도서명', '저자', 'ISBN'], ignore_index=True)

# 결과 확인하기
del_duplicates_book
```

## 공공 데이터를 이용하여 중복되는 데이터를 그룹별로 모으기

```python
### 중복 여부를 확인하여 그룹별로 묶을 컬럼과 '대출건수'컬럼 --> fancy 인덱싱 --> 별도의 DataFrame 생성
loan_df = ns_book[['도서명', '저자', 'ISBN', '대출건수']]

# 결과 확인하기
loan_df
```

```python
### groupby() 함수 적용하기

'''
# 매개변수
1. by
2. dropna=True
3. groupby().sum()
'''

# groupby() 함수 적용
grouped_df = loan_df.groupby(by=['도서명', '저자', 'ISBN'])
print(f'통계 함수를 적용하지 않은 경우의 결과 : \n{grouped_df}')

print('-'*80)

# sum() 사용 --> '대출건수' 컬럼에 적용
loan_count = grouped_df.sum()

# 결과 확인하기
print(f'통계 함수를 적용한 경우의 결과 :')
loan_count
```

```python
### groupby() 함수 적용 + 값 크기 순으로 정렬

# sort_values(by, ascending)

# 오름차순 정렬 --> 기본 값
print(loan_count.sort_values(by='대출건수', ascending=True))

print('-'*80)

# 내림차순 정렬 --> ascending=False
print(loan_count.sort_values(by='대출건수', ascending=False))
```

```python
### 2. boolean 인덱싱 --> 조건에 해당하는 값 추출

# file_name 설정
file_name = '시험점수.csv'

# csv file --> DataFrame 자료형 생성
score_df = pd.read_csv(file_name, encoding='cp949')

# 생성된 데이터 확인
print(score_df)
print('-'*80)

'''
문제 1. 전체 학년 중 3학년 학생의 데이터 추출

문제 2. 국어 점수가 70점 이상인 학생의 데이터 추출

문제 3. 성별이 여자인 데이터 추출

문제 4. 3학년 학생이고 국어 점수가 70점 이상이며 성별이 여자인 데이터 추출
'''

### 풀이 1.
three = (score_df['학년'] == 3)
three_df = score_df.loc[three, :]
print(three_df)
print('-'*80)

### 풀이 2.
seventy = (score_df['국어'] >= 70)
seventy_df = score_df.loc[seventy, :]
print(seventy_df)
print('-'*80)

### 풀이 3.
female = (score_df['성별'] == '여자')
female_df = score_df.loc[female, :]
print(female_df)
print('-'*80)

### 풀이 4.
res = three & seventy & female
res_df = score_df.loc[res, :]
res_df
```
