```python
### random 모듈

import random

# 지정한 범위내의 정수 1개 --> 무작위로 생성
int_value = random.randint(0, 100)
print(int_value)

print('-'*80)

# 0과 1사이의 실수 1개 --> 무작위로 생성
float_value = random.random()
print(float_value)
```

```python
### datetime 모듈

import datetime

# 현재 시간 정보 추출
present_time = datetime.datetime.now()
print(f'현재 시간 정보 : {present_time}')

print('-'*80)

# 년(year) 추출
print(present_time.year)

# 월(month) 추출
print(present_time.month)

# 날짜(day) 추출
print(present_time.day)
```

```python
### urllib(라이브러리)

import urllib.request
url = "http://www.yes24.com/Product/Goods/91065309"
response = urllib.request.urlopen(url)
print(response.read())
```
