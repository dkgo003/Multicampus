## 파이썬으로 API 호출하기

```python
### 필요한 모듈
import requests
```

```python
### HTTP GET 방식으로 API 호출 --> 호출 URL 완성

'''
1. format : format=json 으로 지정, 지정하지 않으면 xml로 받는다
2. startDt : 검색 시작 일자, 2022년 12월 1일로 지정
3. endDt : 검색 종료 일자, 2023년 2월 28 일로 지정
4. age : 연령대, 20으로 지정, ‘20대’ 대출 도서를 조회
5. authKey : 인증키, 인증키를 발급받아 API호출 URL 문자열 맨 끝에 추가
'''

url = 'http://data4library.kr/api/loanItemSrch?format=json&startDt=2022-12-01&endDt=2023-02-28&age=20&authKey=3c7ad329e17f810c4f96c668bb55c4e03738ee7042c2bf2e9652b5fe5233d2f7'
```

```python
### 인증키가 추가된 url을 통해서 도서관 정보나루 사이트로부터 20대의 인기대출도서정보 요청하고 결과 값 가져오기

response = requests.get(url)
```

```python
### text 변수를 통해서 결과 값을 출력하기
data = response.text
print(f'API 호출의 결과 값의 자료형 : {type(data)}')
print('-'*80)
print(f'결과값 : {data}')
```

## 파이썬으로 JSON 데이터 다루기

```python
### 필요한 모듈 임폴트하기
import json
```

```python
### json 문자열 생성 : 바깥쪽 '' + 안쪽 ""
book_string = '{"name":"달러구트 꿈 백화점", "작가":"이미예"}'

# 자료형 확인
print(f'생성된 json 문자열의 자료형 : {type(book_string)}')
```

```python
### json 문자열 --> python dict 자료형으로 변환 --> json.loads() 사용
python_dict = json.loads(book_string)

# 변환된 결과 --> 자료형 확인하기
print(type(python_dict))
print('-'*80)
print(f'변환의 결과 값 : \n{python_dict}')

print('-'*80)

# key --> value 검색 가능
book_name = python_dict['name']
print(f'검색의 결과 : {book_name}')
```

```python
### API 호출을 통해서 획득한 json 문자열 데이터를 python dict 자료형으로 변환하기
api_dict = json.loads(data)
print(f'변환 결과 : \n{api_dict}')

print('-'*80)

# 키로 검색 가능
print(api_dict['response'])
```

```python
### python dict 자료형 --> json 문자열로 변환 --> json.dumps() 사용
json_string = json.dumps(python_dict, ensure_ascii=False)
print(json_string)
```

## 도서 정보 추출하기

```python
### API --> json 문자열 data --> json.loads() --> python dict

# python dict로 변환된 데이터
print(api_dict)

print('-'*80)

# key 확인하기
keys = api_dict.keys()
print(keys)
```

```python
### key 'response' --> value 확인하기
values = api_dict['response']
print(values)
```

```python
### values --> python dict --> key 확인
keys2 = api_dict['response'].keys()
print(keys2)
```

```python
### 도서 정보 --> docs

# docs 정보 추출하기
values_docs = api_dict['response']['docs']

# 결과 확인하기
print(values_docs)
```

```python
### 전체 도서 정보 추출하기(1)

# values_docs --> 성분 원소의 개수 확인
print(len(values_docs))

print('-'*80)

books = []

for i in api_dict['response']['docs']:
    books.append(i['doc'])

# 결과 확인하기
print(books)
```

```python
### 전체 도서 정보 추출하기(2)

# 리스트 컴프리헨션으로 작성
books = [i['doc'] for i in api_dict['response']['docs']]

# 결과 확인하기
print(books)
```

```python
print(books[0])

print('-'*80)

mydict = books[0]
myindex = [0]

df1 = pd.DataFrame(mydict, myindex)
print(df1)

print('-'*80)

df1 # 1건만 이렇게 출력가능
```

```python
### 도서 전체 정보 --> pandas DataFrame 자료형으로 변환
df = pd.DataFrame(data=books)

# 결과 확인하기
df
```

## 웹 스크래핑을 이용한 데이터 수집

```python
### 저장된 json file --> pd.read_json() --> DataFrame 자료형으로 변환

# 저장된 파일 경로 설정
file_name = 'best_book.json'

# pd.read_json() 사용
df_books = pd.read_json(file_name)

# 결과 확인하기
df_books
```

```python
### loc 인덱서 사용 --> 'no'컬럼 부터 'isbn13' --> 새로운 DataFrame 생성하기
books = df_books.loc[:, 'no':'isbn13']

# 결과 확인하기
books
```

```python
### dataframe으로부터 달러구트 꿈 백화점 'isbn13' 정보 추출하기
isbn = df_books.loc[0,'isbn13']
print(isbn)
```

```python
### requests.get(url) 함수 이용 --> 도서 검색 결과 웹 페이지 HTML 정보 획득

# yes24에서 도서 검색을 위한 호출 url 작성
# isbn = isbn
url = f'http://www.yes24.com/Product/Search?domain=BOOK&query={isbn}'

# 정보 요청 / 수령
response = requests.get(url)
```

```python
### 획득한 HTML 정보 확인하기
print(response.text)
```

### HTML 에서 데이터 추출하기 : BeautifulSoup

```python
### 필요한 라이브러리 임폴트
from bs4 import BeautifulSoup

# BeautifulSoup 생성 함수 호출
soup = BeautifulSoup(response.text, 'lxml')
```

```python
### 상세 페이지로 연결되는 링크에 관한 정보 가져오기(1)
link = soup.find('a', {'class':'gd_name'})

# 결과 확인하기
print(link)
```

```python
### 상세 페이지로 연결되는 링크에 관한 정보 가져오기(2)
href = link['href']
print(href)
print(f'링크 정보 자료형 : {type(href)}')
```

```python
### '달러구트 꿈 백화점' 상세 페이지 가져오기

# 호출 url 작성
url = 'http://www.yes24.com'+href
# print(url)

# requests.get(url) --> 상세 페이지 html 정보 가져오기
response = requests.get(url)

# 결과 확인하기
print(response.text)
```

```python
### 품목 정보가 들어있는 태그 정보 가져오기
soup = BeautifulSoup(response.text, 'lxml')
detail = soup.find('div', {'id':'infoset_specific'})

# 결과 확인하기
print(detail)
```

```python
### 테이블 태그 --> tr 태그 전체를 가져오기
tr_list = detail.find_all('tr')
print(tr_list)
```

```python
### 태그 안의 텍스트 가져오기 --> get_text() 이용

for tr in tr_list:
    print(tr)
    if "쪽" in tr.find('td').get_text():
        page = tr.find('td').get_text()
        break
    print('-'*80)
print('-'*80)
print(page)
print('-'*80)

# 자료형 확인하기
print(f'page 변수의 자료형 : {type(page)}')
```

```python
### 쪽수 정보 가져오기
page_list = page.split()
print(page_list)

print('-'*80)

# 쪽수 정보
info = page_list[0]
print(info)
```
