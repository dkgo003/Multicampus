## if 조건문

```python
### if 조건문

money = 11000

if money >= 10000:
    print("택시를 타고 집에 간다")
```

```python
### 조건문 if else 실습

money = 11000

if money >= 15000:
    print("택시를 타고 집에 간다")
else:
    print("걸어 간다")
```

```python
### 조건문 if elif else 실습

money = 1200

if money >= 15000:
    print("택시를 타고 집에 간다")
elif 1500 <= money < 15000:
    print("광역 버스를 타고 집에 간다")
else:
    print("걸어 간다")
```

```python
### 어떤 종류의 학생인지 맞히기

print("당신이 태어난 연도를 입력하세요")
birth_year = input()
age = 2023 - int(birth_year) + 1

if  age <= 26 and age >= 20:
    print("대학생")
elif age < 20 and age >= 17:
    print("고등학생")
elif age < 17 and age >= 14:
    print("중학생")
elif age < 14 and age >= 8:
    print("초등학생")
else:
    print("학생이 아닙니다.")
```

## For문

```python
### for문 예시

for looper in [1, 2, 3, 4, 5]:
    print(looper)
    print("Hello")
```

```python
### rnage(시작값, 종료값, 증분(step=1))

for looper in range(5, 100, 4):
    print(looper)
```

```python
### 문자열을 이용

for char in 'abcdef':
    print(char)
    print("안녕")
```

```python
### 문자열로 이루어진 리스트

for i in ['americano', 'latte', 'frappuccino']:
    print(i)
```

```python
### 1부터 9까지 2씩 증가시키는 for문

for i in range(1, 10, 2):
    print(i)
```

```python
### 10부터 2까지 1씩 감소시키는 for문

for i in range(10, 1, -1):
    print(i)
```

## While문

```python
### while문

i = 1
while i < 10:
    print(i)
    i += 1
```

## 반복문의 제어

```python
### break문

for i in range(10):
    if i == 5:
        break
    print(i)
print("End of Program")
```

```python
### continue문

for i in range(10):
    if i == 5:
        continue
    print(i)
print("End of Program")
```

```python
### else문

for i in range(10):
    print(i)
else:
    print("End of Program")
```

```python
### 구구단 계산기

print("구구단 몇 단을 계산할까?")
num = int(input())
print(f"구구단 {num}단을 계산한다.")

for i in range(1, 10):
    res = num * i
    print(num, "x", i, "=", res)
```

## 조건문과 반복문 실습

```python
### 문자열 역순 출력

char = "hello"

reverse_char = ' ' # 스페이스 한칸을 쓰는 이유는 빈칸이라는 것을 표시하기위해
for i in char:
    reverse_char = i + reverse_char
    print(i)
print(reverse_char)
```

```python
### 십진수를 이진수로 변환

num = 84

res = ' '
while num > 0:
    twonum = num % 2
    num = num // 2
    res = str(twonum) + res
    print(res) # 과정 보기 위해서
print(res)
```
