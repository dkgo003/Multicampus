## 한글 텍스트 '형태소'단위 토큰화 실습

```python
### konlpy, Mecab 형태소 분석기 설치 스크립트 실행
!curl -s https://raw.githubusercontent.com/teddylee777/machine-learning/master/99-Misc/01-Colab/mecab-colab.sh | bash
```

```python
### 필요한 라이브러리 임폴트
from konlpy.tag import Okt, Mecab

# 형태소 분석기 객체 생성
okt = Okt()
mecab = Mecab()
```

```python
### Okt 형태소 분석기 실습

# 실습 문장 생성
text = "한글 형태소 분석기로 테스트를 해보았습니다. 설치 및 동작이 잘 됩니다."

# 문장 --> 형태소 추출 및 분할(토큰화) --> morphs(stem, norm)
morphs_okt = okt.morphs(text)
morphs_stem_okt = okt.morphs(text, stem=True)

# 비교(1) : 공백 단위로 분할 --> split()
splited = text.split()

# 비교(2) : word_tokenize() 사용
tokenized_text = word_tokenize(text)

# 결과 확인하기
print(f'Okt 형태소 분석기를 이용하여 형태소 분석을 한 결과 : \n{morphs_okt}')
print('-'*80)
print(f'Okt 형태소 분석기(어간)를 이용하여 형태소 분석을 한 결과 : \n{morphs_stem_okt}')
print('-'*80)
print(f'공백 단위로 분할한 결과 : \n{splited}')
print('-'*80)
print(f'word_tokenize 함수를 이용하여 토큰화 한 결과 : \n{tokenized_text}')
```

```python
### 형태소와 품사를 동시에 추출 --> 품사 tagging --> pos() 함수 사용
pos_okt = okt.pos(text)
pos_stem_okt = okt.pos(text, stem=True)

# 결과 확인하기
print(f'형태소와 품사를 동시에 추출한 결과 : \n{pos_okt}')
print('-'*80)
print(f'형태소(어간)와 품사를 동시에 추출한 결과 :\n{pos_stem_okt}')
```

```python
### Mecab 형태소 분석기 실습

# 실습 문장
text = "한글 형태소 분석기로 테스트를 해보았습니다. 설치 및 동작이 잘 됩니다."

# 문장으로부터 형태소 추출 --> morphs() 함수 사용 --> 형태소로 분할
morphs_mecab = mecab.morphs(text)

# 결과 확인하기
print(f'mecab 형태소 분석기를 이용한 형태소 분석의 결과 : \n{morphs_mecab}')
print('-'*80)
print(f'okt 형태소 분석기를 이용한 형태소 분석의 결과 : \n{morphs_okt}')
```

```python
### 형태소와 품사를 동시에 추출 --> pos() 함수 사용
pos_mecab = mecab.pos(text)

# 결과 확인하기
print(f'mecab 형태소 분석기로 형태소와 품사를 동시에 분석한 결과 : \n{pos_mecab}')
print('-'*80)
print(f'okt 형태소 분석기로 형태소와 품사를 동시에 분석한 결과 : \n{pos_okt}')
```

```python
### 문장에서 명사, 동사, 형용사만 추출

# 품사가 태깅된 데이터 형태 확인
data = pos_stem_okt[0] #('text', 'pos')
morpheme = data[0] # 형태소만 추출
pos = data[1] # 품사만 추출
print(f'품사가 태깅된 데이터 1개 = {data}')
print(f'형태소만 추출 = {data[0]}')
print(f'품사만 추출 = {data[1]}')

print('-'*80)

# 추출 결과를 저장할 리스트 생성
okt_morphs = []
mecab_morphs = []

# Okt 형태소 분석기를 이용하여 명사, 동사, 형용사 추출
for data in pos_stem_okt:
    if 'Noun' in data or 'Verb' in data or 'Adjective' in data:
        okt_morphs.append(data[0])

# Mecab 형태소 분석기를 이용하여 명사, 동사, 형용사 추출
for data in pos_mecab:
    if "NNG" in data[1].split("+") or "NNP" in data[1].split("+") or "VV" in data[1].split("+") or "VA" in data[1].split("+"):
        mecab_morphs.append(data[0])

# 결과 확인하기
print(f'Okt 형태소 분석기를 이용하여 명사, 동사, 형용사만 추출한 결과 : \n{okt_morphs}')
print('-'*80)
print(f'Mecab 형태소 분석기를 이용하여 명사, 동사, 형용사만 추출한 결과 : \n{mecab_morphs}')
```

## ByteLevelBPE Tokenizer 실습

```python
'''
실습 개요
- naver sentiment movie corpus(nsmc) 데이터를 가지고 BPE 단어 사전을 구축한다
- BPE(Byte Pair Encoding)이 끝나고 모델을 저장하면, 단어 사전 파일(vocab.json)과
  바이그램 쌍의 병합 우선 순위를 나타내는 파일 (merges.txt)이 생성된다
- 결과물로 생성되는 vocab.json과 merges.txt는 bpe 폴더에 저장되도록 한다.
'''
```

```python
### 실습에 필요한 라이브러리 설치

# HuggingFace --> tokenizers 라이브러리 설치
!pip install tokenizers
```

```python
### ByteLevelBPE Tokenizer 임폴트, 객체 생성
from tokenizers import ByteLevelBPETokenizer

tokenizer = ByteLevelBPETokenizer()
```

```python
### 텍스트 데이터 --> 토큰화 --> 단어 사전 생성(word : id)

# 구글 서버와 내 드라이브 연결
from google.colab import drive
drive.mount('/content/drive')

# 현재 작업 디렉토리 변경
%cd '/content/drive/MyDrive/KDT/비정형텍스트분석'

# 파일 경로 설정
train_path = 'nsmc_train.txt'
test_path = 'nsmc_test.txt'

tokenizer.train(
    files=[train_path, test_path],
    vocab_size=1000,
    min_frequency=1
)
```

```python
### get_vocab() --> 생성된 단어 사전 불러오기
vocab = tokenizer.get_vocab()

# 단어 사전의 자료형 확인하기
print(f'단어 사전의 자료형 = {type(vocab)}')

print('-'*80)

# 단어 사전 출력하기
print(vocab)
```

```python
### BPE(ByteLevelBPETokenizer)적용하기

# 텍스트 문장 생성
text = "너와 함께한 시간 모두 눈부셨다. 날이 좋아서 날이 좋지 않아서 날이 적당해서 모든 날이 좋았다."

# 토큰화 + 단어 사전을 참조 --> 문장 숫자화 --> encode() 적용
encoded = tokenizer.encode(text)

# 결과 확인하기
print(encoded)
```

```python
### Encoding 결과 --> 속성 확인하기
print(f'ids = \n{encoded.ids}')
print('-'*80)
print(f'tokens = \n{encoded.tokens}')
```

```python
### 한글화 --> decode() 적용하기
decoded = tokenizer.decode(encoded.ids)
print(decoded)
```

```python
### WordPieceTokenizer 학습 --> 단어 사전 생성

wordpiece_tokenizer.train(
    files=[train_path, test_path],
    vocab_size=1000,
    min_frequency=1
)
```

```python
### get_vocab() 함수 --> 생성된 단어 사전 불러오기
wordpiece_vocab = wordpiece_tokenizer.get_vocab()

# 단어 사전의 자료형 확인하기
print(f'생성된 단어 사전의 자료형 = {type(wordpiece_vocab)}')

print('-'*80)

# 단어 사전 출력
print(wordpiece_vocab)
```

```python
### WordPieceTokenizer 적용하기

# 텍스트 문장 생성
text = "너와 함께한 시간 모두 눈부셨다. 날이 좋아서 날이 좋지 않아서 날이 적당해서 모든 날이 좋았다."

# 토큰화 + 단어 사전을 참조 --> 문장 숫자화 --> encode() 적용
wordpiece_encoded = wordpiece_tokenizer.encode(text)

# 결과 확인하기
print(wordpiece_encoded)
```

```python
### Encoding 속성 확인하기
print(f'ids = {wordpiece_encoded.ids}')
print('-'*80)
print(f'tokens = {wordpiece_encoded.tokens}')
```

```python
### decode() --> 한글화
wordpiece_decoded = wordpiece_tokenizer.decode(wordpiece_encoded.ids)
print(wordpiece_decoded)
```
