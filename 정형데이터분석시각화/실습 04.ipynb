### 이상치 처리하기

```python
### 이상치가 있는 데이터 불러오기

# file 경로 설정
file_name = 'weather.csv'

# read_csv() 실행
df_weather = pd.read_csv(file_name)

# 결과 확인하기 --> info()
print("기본정보 : ")
df_weather.info()
print('-'*80)
print(f'데이터의 모양 = {df_weather.shape}')
print('-'*80)
print(f'전체 데이터 출력 : \n{df_weather}')
```

### 3/31 실습

```python
### 각 컬럼 별 4분위 수 구하기

'''
df.col.quantile(q=)
'''

# 1/4 분위수 구하기
q1 = df_weather.quantile(q=0.25)
print(f'1사분위수 = \n{q1}')

print('-'*80)

# 2/4 분위수 구하기
q2 = df_weather.quantile(q=0.5)
print(f'2사분위수 = \n{q2}')

print('-'*80)

# 3/4 분위수 구하기
q3 = df_weather.quantile(q=0.75)
print(f'3사분위수 = \n{q3}')
```

```python
### 각 컬럼 별 IQR 구하기

'''
IQR = q3 - q1
'''

IQR = q3 - q1
print(IQR)
```

```python
### 각 컬럼 별 정상 범위 데이터의 최소값 / 최대값 구하기

'''
1. min = q1 - (1.5 * IQR)
2. max = q3 + (1.5 * IQR)
3. 정상 데이터 : (min, max)
'''

# 최소값 구하기
min = q1 - (1.5 * IQR)
print(f'정상 데이터의 최소값 : \n{min}')
print('-'*80)

# 최대값 구하기
max = q3 + (1.5 * IQR)
print(f'정상 데이터의 최대값 : \n{max}')
print('-'*80)
```

```python
### 이상치 제거 함수 실행

def remove_outlier(df, col):
    data = df[col]
    q1 = data.quantile(q=0.25)
    q3 = data.quantile(q=0.75)
    iqr = q3 - q1
    iqr = iqr * 1.5
    min = q1 - iqr
    max = q3 + iqr
    # boolean 인덱싱 이용
    outlier_index = data[(data<min) | (data>max)].index
    df.drop(index=outlier_index, axis=0, inplace=True)
    print(f'이상치 제거 후 데이터의 모양 = {df.shape}')
    return df
```

```python
### 이상치 제거 함수 실행

# temp 컬럼 실행
remove_outlier(df_weather, 'temp')
print('-'*80)

# hum 컬럼 실행
remove_outlier(df_weather, 'hum')
print('-'*80)

# atm 컬럼 실행
remove_outlier(df_weather, 'atm')
print('-'*80)

# speed 컬럼 실행
remove_outlier(df_weather, 'speed')
print('-'*80)
```

### Binning(구간화)

```python
'''
### pd.cut(x, bins, labels)
'''

# 데이터 설정
ages = [1, 10, 15, 13, 21, 23 ,37, 31, 43, 80, 61, 20, 41, 32, 100]
bins = [0, 15, 25, 35, 60, 100]
labels = ['어린이', '청년', '장년', '중년', '노년']

# pd.cut() 적용
cuts = pd.cut(ages, bins, labels)

# 결과 확인하기
print(f'전체 정보 확인 : \n{cuts}')
print('-'*80)
print(f'범주 정보 확인 : \n{cuts.categories}')
```
