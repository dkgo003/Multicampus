## 인덱싱 / 슬라이싱

```python
### 인덱싱
mystring = 'Python is my life.'
print(mystring)

print('-'*80)

# 0번째 문자 --> 추출
print(mystring[0])

print('-'*80)

# 마지막 문자 --> 추출
print(mystring[-1])
```

```python
### 슬라이싱
mystring = 'Python is my life.'
print(mystring)

print('-'*80)

# 전체 범위 지정
print(mystring[:])

print('-'*80)

# 전체 + 역순 --> [::-1]
print(mystring[::-1])

print('-'*80)

# 첫문자부터 2칸씩 간격으로 추출
print(mystring[::2])
```

## 문자열의 덧셈과 곱셈

```python
### 문자열의 덧셈 --> 문자열 데이터의 연결
string_a = '안녕하세요'
string_b = '반갑습니다.'

res = string_a + " " + string_b
print(res)
```

```python
### 문자열의 곱셈 --> 반복적인 연결
res = string_a * 3
print(res)
```

## 자료형 변환

```python
### 문자열 --> 리스트
mystring = '가나다라마'
str_list = list(mystring)
print(str_list)
```

```python
### 문자열 --> 튜플
mystring = 'abcdef'
str_tuple = tuple(mystring)
print(str_tuple)
```

```python
### 문자열 --> 집합(set)
mystring = 'AAACCCBBB'
# 중복 제거
str_set = set(mystring)
print(str_set)
# set --> list
mylist = list(str_set)
print(mylist)
```

## 함수

```python
### split('기준')
mystring = 'This is my life.'
str_list = mystring.split()
print(str_list)
```

```python
### ''.join(리스트[word1, word2, ....])

# str_list 이용 -> 복원
mystring = ' '.join(str_list)
print(mystring)

print('-'*80)

# '/' 첨가 복원
mystring1 = '/'.join(str_list)
print(mystring1)

print('-'*80)

# '-'첨가 복원
mystring2 = '-'.join(str_list)
print(mystring2)
```

### quiz(1)

```python
'''
sample = 'abc-def.789/100'

결과1 : ['abc', 'def.789/100']
결과2 : 789추출
'''

sample = 'abc-def.789/100'
res = sample.split('-')
print(res)

print('-'*80)

sample1 = res[1]
res1 = sample1.split('.')
print(res1)

print('-'*80)

sample2 = res1[1]
res2 = sample2.split('/')
print(res2)

print('-'*80)

print(res2[0])

print('-'*80)

# 압축
result = sample.split('-')[1].split('.')[1].split('/')[0]
print(result)
```
