## 람다 함수

```python
### [코드 9-1]

def f(x, y):
    return x + y

print(f(1, 4))
```

```python
### [코드 9-2]

f = lambda x, y: x + y
print(f(1, 4))
```

```python
f = lambda x, y: x + y
print(f(1, 4))

f = lambda x: x ** 2
print(f(3))

f = lambda x: x / 2
print(f(3))

f = lambda x: x / 2
print(f(3, 5))
```

```python
### 리스트 컴프리헨션으로 변경한 코드

# 리스트 자료형 데이터 생성
ex = [1, 2, 3, 4, 5]

# 일반적인 for 문

res = []

for i in ex:
    res.append(i**2)

print(res)

# 리스트 컴프리헨션 구문 변경
res = [i**2 for i in ex]
print(res)
```

## 별표의 활용

```python
### 가변 인수

def asterisk_test(a, *args):
    print(a, args)
    print(type(args))

asterisk_test(1, 2, 3, 4, 5, 6)
```

```python
### 키워드 가변 인수

def asterisk_test(a, **kargs):
    print(a, kargs)
    print(type(kargs))

asterisk_test(1, b=2, c=3, d=4, e=5, f=6)
```

## 별표의 언패킹 기능

```python
def asterisk_test(a, args):
    print(a)
    print(f'args = {args}')
    print(*args)
    print(type(args))

asterisk_test(1, (2, 3, 4, 5, 6))
```

```python
def asterisk_test(a, *args):
    print(a, args)
    print(type(args))

asterisk_test(1, *(2, 3, 4, 5, 6))
```

```python
a, b, c = ([1, 2], [3, 4] ,[5, 6])
print(a, b, c)

data = ([1, 2], [3, 4] ,[5, 6])
print(*data)
```

```python
for data in zip(*[[1,2,3], [3,4,5], [5,6,9]]):
    print(data)
    print(type(data))
```

```python
def asterisk_test(a, b, c, d):
    print(a, b, c, d)

data = {"b":1, "c":5, "d":2}
print(data)
asterisk_test(10, **data)
```

## 선형대수학

### 벡터(vector)

```python
### [그림 9-7]의 수식을 코드로 작성

# 1차원 벡터 --> 데이터 생성
u = [2, 2]
v = [2, 3]
z = [3, 5]

result =[]

# 성분 원소의 개수 만큼만 연산 수행 --> range(len(u))
for i in range(len(u)):
    result.append(u[i] + v[i] + z[i])

print(result)
```

```python
### 파이썬 스타일 코드를 이용

u = [2, 2]
v = [2, 3]
z = [3, 5]

result = [sum(t) for t in zip(u, v, z)]
print(result)

# 확인
[t for t in zip(u, v, z)]
```

```python
### 별표를 사용한 함수화

u = [2, 2]
v = [2, 3]
z = [3, 5]

def vector_addition(*args):
    print(f'args = {args}')
    return [sum(t) for t in zip(*args)]

result = vector_addition(u ,v, z)
print(result)

print('-'*80)

# 이차원 리스트 형식의 데이터 생성 --> '행렬'
row_vector = [[2, 2], [2, 3], [3, 5]]

result = vector_addition(*row_vector)
print(result)
```

```python
###  [그림 9-8]의 수식을 코드로 작성

# 1차원 리스트 자료형 데이터 생성
u = [1, 2, 3]
v = [4, 4, 4]

# 0차원 스칼라 데이터 생성
alpha = 2

# 연산을 위한 수식 작성
result = [alpha * sum(t) for t in zip(u, v)]
print(result)
```

### 행렬(matrix)

```python
### 행렬의 연산을 파이썬 스타일 코드를 사용하여 표현하기

matrix_a = [[3, 6], [4, 5]]
matrix_b = [[5, 8], [6, 7]]
result = [[sum(row) for row in zip(*t)] for t in zip(matrix_a, matrix_b)]

print(result)
```

```python
### 전치행렬(transpose matrix)

matrix_a = [[1, 2, 3], [4, 5, 6]]
result = [[element for element in t] for t in zip(*matrix_a)]
print(result)

result = [t for t in zip(*matrix_a)]
print(result)
```

```python
matrix_a = [[1, 1, 2], [2, 1, 1]]
matrix_b = [[1, 1], [2, 1], [1, 3]]
result = [[sum(a * b for a, b in zip(row_a, column_b)) for column_b in zip(*matrix_b)] for row_a in matrix_a]
print(result)
```
