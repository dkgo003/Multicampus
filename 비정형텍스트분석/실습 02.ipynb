### CountVectorizer 실습

```python
### 필요한 라이브러리 임폴트
from sklearn.feature_extraction.text import CountVectorizer
```

```python
### CountVectorizer 객체 생성
ko_count = CountVectorizer()
en_count = CountVectorizer()
```

```python
### 한글 실습용 텍스트 데이터 생성 및 단어 사전 생성하기

# 실습용 텍스트 데이터 생성
ko_text = ['나는 배가 고프다', '내일 점심 뭐먹지', '내일 공부 해야겠다', '점심 먹고 공부해야지']

# fit() 함수 이용 --> 자동으로 토큰화, 단어 사전 생성
ko_count.fit(ko_text)

# 생성된 단어 사전 확인하기
ko_vocab = ko_count.vocabulary_
print(ko_vocab)
```

```python
### 영어 실습용 텍스트 데이터 생성 및 단어 사전 생성하기

# 영어 텍스트 데이터 생성
en_text = ['You already know what The Present is', 'You already know where to find it', 'And you already know how it can make you happy and successful', 'You knew it best when you were younger', 'You have simply forgotten']

# fit() 함수 --> 자동으로 단어 사전 생성 (순서 : 알파벳 순서)
en_count.fit(en_text)

# 생성된 단어 사전 확인하기
en_vocab = en_count.vocabulary_
print(en_vocab)

print('-'*80)

# 생성된 단어 사전 --> 정렬
sorted_en_vocab = sorted(en_vocab.items())
print(sorted_en_vocab)
```

## One-Hot Encoding 실습

```python
### 필요한 라이브러리 임폴트
import tensorflow as tf
```

```python
### Tokenizer 객체 생성
tokenizer = tf.keras.preprocessing.text.Tokenizer()
```

```python
### 실습용 텍스트 데이터 생성 및 단어 사전 생성하기

# 실습용 데이터
ko_text = ['나는 배가 고프다', '내일 점심 뭐먹지', '내일 공부 해야겠다', '점심 먹고 공부해야지']

# fit_on_texts() 함수 이용 --> 단어 사전 생성
tokenizer.fit_on_texts(ko_text)

# 단어 사전 확인하기 --> word_index 이용
word_to_index = tokenizer.word_index
print(f'단어 사전(단어 : 인덱스) : \n{word_to_index}')

print('-'*80)

# 단어들의 빈도수 확인 --> word_counts 이용
counts = tokenizer.word_counts
print(f'단어들의 빈도수 : \n{counts}')
```

```python
### 문장의 행렬화

# 실습용 문장 생성 --> 리스트 자료형(1차원 데이터)
sentence = [ko_text[3]]

# texts_to_sequences() 함수 사용 --> 정수 인코딩 생성
encoded = tokenizer.texts_to_sequences(sentence)
print(f'정수 인코딩의 결과 : {encoded}')
```

```python
### to_categorical() 함수 --> (단어 : One-Hot Encoding 벡터) --> 문장 --> 원-핫 인코딩 행렬화
one_hot = tf.keras.utils.to_categorical(encoded)

# 결과 확인하기
print(f'원-핫 인코딩의 결과 : \n{one_hot}')
```

## 텍스트 유사도 측정 - 코사인 유사도

```python
### 실습 내용

'''
1. 두 개의 문장을 TfidfVecterizer를 이용해서 행렬로 나타낸다.
2. 코사인 유사도 공식을 사용하여 두 문장 간의 유사도를 측정
'''

# 필요한 라이브러리 임폴트
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
### 필요한 라이브러리 임폴트
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
### 데이터 불러오기

# 구글 서버와 내 드라이브 연결
from google.colab import drive
drive.mount('/content/drive')

# 작업 경로 변경하기
%cd '/content/drive/MyDrive/KDT/비정형텍스트분석'

# 파일 경로 설정
file_name = 'movies_metadata.csv'

# pd.read_csv() --> dataframe 자료형 생성
df_movies = pd.read_csv(file_name)

# 모든 컬럼 표시 --> option 설정
pd.options.display.max_columns = None

# 결과 확인하기
print(df_movies)
```

```python
### 줄거리, 제목 컬럼을 기준으로 누락 데이터 확인하고 제거 --> 행 제거

# 줄거리(overview) 컬럼에 있는 누락 데이터 확인
nulls_overview = df_movies.overview.isnull().sum()
print(f'줄거리 컬럼의 누락 데이터의 수 = {nulls_overview}')

print('-'*80)

# 제목(title) 컬럼에 있는 누락 데이터 확인
nulls_title = df_movies.title.isnull().sum()
print(f'제목 컬럼의 누락 데이터의 수 = {nulls_title}')

print('-'*80)

# 누락 데이터 제거 --> dropna(subset)
df_movies.dropna(subset=['overview', 'title'], inplace=True)

# 결과 확인하기 --> info() 이용
df_movies.info()
```

```python
### TfidfVectorizer 객체 생성

# stop_words='english' 추가 --> 불용어 제거
tfidf = TfidfVectorizer(stop_words='english')
```

```python
### fit() 함수를 이용 --> overview 컬럼에 대해서 단어 사전 생성

# data의 수 --> 10000개로 축소 --> iloc 인덱서를 이용 --> 슬라이싱
data = df.iloc[:10000, :]

tfidf.fit(data.overview)

# 생성된 단어 사전의 수 확인하기
vocab = tfidf.vocabulary_
print(f' 단어 사전에 들어있는 단어의 수 = {len(vocab)}')
print('-'*80)
print(f'단어 사전 확인 : \n{vocab}')
```

```python
### transform().toarray() --> 문장 행렬화
overview_matrix = tfidf.transform(data.overview).toarray()

# 결과 확인하기
print(f'overview 컬럼의 각 문장을 행렬로 변환한 결과(모양) = {overview_matrix.shape}')
```

```python
### 코사인 유사도 --> linear_kernel

# linear_kernel() --> 10000개 영화의 줄거리 각각에 대해서 cosine 유사도 구하기
similarity = linear_kernel(overview_matrix, overview_matrix)

# 결과 확인하기
print(f'유사도 측정의 결과 : \n{similarity}')
```

```python
### 전체 영화에 대한 유사도 행렬 --> 특정 영화와 모든 영화와의 유사도 측정 결과 추출
similarity[10]
```

```python
### 코사인 유사도 --> 데이터프레임 생성

# 컬럼과 인덱스 설정
columns = data.title.values
index = data.title.values

# 데이터 프레임 생성
df_sim = pd.DataFrame(data=similarity, columns=columns, index=index)

# 결과 확인하기
df_sim
```

```python
### 특정 영화 기준 --> 줄거리가 유사한 영화 추천 함수 생성(1)

'''
1. 유사도 크기 순으로 추출
2. 해당 영화의 유사도 제거
3. 특정 영화와 다른 영화와의 유사도 크기 순으로 10개만 추출
'''
top10 = df_sim['Toy Story'].sort_values(ascending=False)[1:11]

# 결과 확인하기
print(f'결과 확인하기 : \n{top10}')
```

```python
### 영화의 title을 인덱스로 하고, 데이터의 인덱스를 값으로 하는 pandas series
extract_index = pd.Series(data=data.index, index=data.title)

# 결과 확인하기
print(f'생성된 Series 자료형 확인 : \n{extract_index}')
```

```python
### extract_index 용도 : 영화의 제목을 입력하면 해당 영화의 인덱스를 추출
idx = extract_index['Toy Story']

# 결과 확인하기
print(f'해당 영화의 인덱스 = {idx}')
```

```python
### 특정 영화 기준 --> 줄거리가 유사한 영화 추천 함수 생성(2)

def recommand(title, simiarity):
    # 선택한 영화의 제목으로부터 해당 영화의 인덱스를 추출
    idx = extract_index[title]
    # 모든 영화에 대해서 해당 영화와의 유사도를 구한다
    sim_scores = [(idx, score) for idx, score in enumerate(similarity[idx])]
    # print(sim_scores)
    # sorted() --> 유사도에 따라서 영화들을 정렬 --> 내림차순 정렬
    sim_scores = sorted(sim_scores, key= lambda x:x[1], reverse=True)
    # print(sim_scores)
    # 해당 영화를 제외하고 유사도가 가장 큰 영화 n개를 추출한다
    sim_scores = sim_scores[1:11]
    # print(sim_scores)
    # 유사도가 가장 큰 10개의 영화의 인덱스를 추출한다
    movie_index = [data[0] for data in sim_scores]
    # index --> title 추출하기
    movie_title = data.title[movie_index].values

    return movie_title
```

```python
### 추천 함수 적용하기
top10 = recommand('GoldenEye', similarity)

# 결과 확인하기
print(top10)
```

```python
### 데이터 전처리 --> 누락 데이터 제거

# 누락 데이터 확인 --> isnull().sum()
num_nulls = df.isnull().sum()
print(num_nulls)

print('-'*80)

# dropna() 사용 --> 누락 데이터가 있는 행 제거
df_review = df.dropna()

# 누락 데이터 제거 후 결과 확인하기
print(f'누락 데이터 제거 후 기본 정보 확인')
df_review.info()
```

```python
### pd.series.str.replace() 함수 --> 한글과 공백 외 문자 제거
df_review.document = df_review.document.str.replace("[^가-횧]", "")

# 결과 확인하기
df_review
```

```python
### 불용어 정의
stopwords = ['의', '가', '이', '은', '들', '는', '좀', '잘', '걍', '과', '도', '를', '자', '에']
```

### Okt 형태소 분석기를 이용한 Word2Vec실습

```python
### Okt 형태소 분석기 --> 토큰화 작업, 불용어 제거

# Okt 형태소 분석기 객체 생성
okt = Okt()

# okt.morphs(stem=True)
tokenized_data1 = []
for review in df_review.document:
    sentence_tokenized = []
    # 토큰화
    tokenized_review = okt.morphs(review, stem=True)
    # 불용어 제거
    for token in tokenized_review:
        if token not in stopwords:
            sentence_tokenized.append(token)
    tokenized_data1.append(sentence_tokenized)
```

```python
### 불용어를 제거하고 토큰화 해준 데이터 --> 5개 확인
tokenized_data1[:5]
```

```python
### CBOW 모델 --> 생성된 단어 사전과 임베딩 벡터 확인하기

'''
1. 임베딩 벡터 생성 --> model.wv 
2. 단어 사전 : model.wv.key_to_index 
3. model.wv['word']를 통해 특정 단어의 embedding vector를 확인할 수 있다
4. model.wv.vectors 속성을 이용하여 생성된 단어 사전의 전체 value(embedding vector)
를 확인할 수 있다.
5. model.wv.similarity(w1=w1, w2=w2) 메서드를 이용하여 단어 간의 유사도를 비교할 수 있다.
6. model.wv.most_similar(word, topn=n) 메서드를 이용하여 특정 단어와 가장 가까운 단어
n개를 추출할 수 있다.

'''

# 단어 사전 확인하기
vocab = loaded_cbow_okt.wv.key_to_index
print(f'단어 사전의 자료형 = {type(vocab)}')
print('-'*80)
print(f'단어 사전 확인 : \n{vocab}')

print('-'*80)

# 단어 사전의 전체 key(token, 형태소) 확인 
tokens = vocab.keys()
print(f'단어 사전에 있는 형태소 : \n{tokens}')
print('-'*80)
print(f'단어 사전의 형태소의 수 = {len(tokens)}')

print('-'*80)

# 특정 단어의 임베딩 벡터 확인
wv = loaded_cbow_okt.wv['영화']
print(f"'영화'라는 형태소에 대한 임베딩 벡터 : \n{wv}")

print('-'*80)

# 두 단어의 유사도 비교 - '황정민', '배우'
similarity = loaded_cbow_okt.wv.similarity("황정민", "배우")
print(similarity)

print('-'*80)

# 특정 단어와 가장 가까운 단어 n개 추출 
top5 = loaded_cbow_okt.wv.most_similar("황정민", topn=5)
print(top5)
```

```python
### skip-gram 모델 --> 생성된 단어 사전과 임베딩 벡터 확인

# 단어 사전 확인
vocab = loaded_sg_okt.wv.key_to_index
print(f'단어 사전의 자료형 : {type(vocab)}')
print('-'*80)
print(f'단어 사전 : {vocab}')

print('-'*80)

# 특정 단어의 embedding vector 확인
wv = loaded_sg_okt.wv['영화']
print(f"'영화'라는 형태소에 대한 임베딩 벡터 : \n{wv}")

print('-'*80)

# 두 단어의 유사도 비교 - '황정민', '배우'
similarity = loaded_sg_okt.wv.similarity("황정민", "배우")
print(similarity)

print('-'*80)

# 특정 단어와 가장 가까운 단어 n개 추출 
top5 = loaded_sg_okt.wv.most_similar("황정민", topn=5)
print(top5)
```

### Word2Vec을 이용한 문장 유사도 측정

```python
### nltk 라이브러리 설치
!pip install nltk
```

```python
### 필요한 라이브러리 임폴트
import nltk
from nltk.tokenize import word_tokenize
from sklearn.metrics.pairwise import cosine_similarity
from gensim.models import Word2Vec
import pandas as pd
```

```python
### 문장 부호 인식하기 위해서 필요한 추가 설치
nltk.download('punkt')
```

```python
### 두 문장간의 유사도 비교 --> 영어 텍스트 데이터 생성
en_sentences = ["What should I do to be a great scientist?", "How can I be a good scientist?"]
print(en_sentences)
```

```python
### 영어 텍스트 --> word --> tokenization

tokenized_data = []

for sentence in en_sentences:
    tokenized_list = word_tokenize(sentence)
    tokenized_data.append(tokenized_list)

# 결과 확인하기
print(tokenized_data)
```

```python
### Word2Vec 실행
model_cbow = Word2Vec(sentences=tokenized_data, vector_size=10, window=5, min_count=1, workers=1, sg=0)
model_sg = Word2Vec(sentences=tokenized_data, vector_size=10, window=5, min_count=1, workers=1, sg=1)
```

```python
### CBOW 모델 --> 생성된 단어 사전 / 임베딩 벡터 확인

# 단어 사전 확인
vocab = model_cbow.wv.key_to_index
print(f'생성된 단어 사전 확인 : {vocab}')

print('-'*80)

# 특정 단어에 대한 임베딩 벡터 확인
wv_cbow = model_cbow.wv['scientist']
print(wv_cbow)
```

```python
### skip-gram 모델 --> 생성된 단어 사전 / 임베딩 벡터 확인

# 단어 사전 확인
vocab = model_sg.wv.key_to_index
print(f'생성된 단어 사전 확인 : {vocab}')

print('-'*80)

# 특정 단어에 대한 임베딩 벡터 확인
wv_sg = model_sg.wv['scientist']
print(wv_sg)
```

```python
### 문장의 행렬화

sentence_matrix1 = []
sentence_matrix2 = []

# 첫번째 문장 --> 행렬화
for token in tokenized_data[0]:
    token_vec = model_sg.wv[token]    
    sentence_matrix1.append(token_vec)

# 두번째 문장 --> 행렬화
for token in tokenized_data[1]:
    token_vec = model_sg.wv[token]
    sentence_matrix2.append(token_vec)

# 결과 확인하기
print(f'첫번째 문장에 대한 행렬 값 : \n{sentence_matrix1}')
print('-'*80)
print(f'두번째 문장에 대한 행렬 값 : \n{sentence_matrix2}')
```

```python
### 코사인 유사도 측정하기
similarity = cosine_similarity(sentence_matrix1, sentence_matrix2)

# 결과 확인하기
print(similarity)
```

```python
### 문장의 행렬화 개선 --> 첫번째 문장 

# 문장 행렬 --> 데이터프레임 생성
df_sent1 = pd.DataFrame(data=sentence_matrix1)

# 생성된 데이터프레임의 결과 확인하기
print(df_sent1)

print('-'*80)

# 개선된 문장 행렬 생성
sent_matrix1 = df_sent1.mean().values.reshape((1, -1))

# 결과 확인하기
print(f'개선된 문장 행렬 생성 결과 : \n{sent_matrix1}')
```

```python
### 문장의 행렬화 개선 --> 두번째 문장 

# 문장 행렬 --> 데이터프레임 생성
df_sent2 = pd.DataFrame(data=sentence_matrix2)

# 생성된 데이터프레임의 결과 확인하기
print(df_sent2)

print('-'*80)

# 개선된 문장 행렬 생성
sent_matrix2 = df_sent2.mean().values.reshape((1, -1))

# 결과 확인하기
print(f'개선된 문장 행렬 생성 결과 : \n{sent_matrix2}')
```

```python
### 코사인 유사도 측정하기
similarity = cosine_similarity(sent_matrix1, sent_matrix2)
print(similarity)
```
