```python
'''
# 문제1

아래의 혈액형 통계 수치를 이용하여 혈액형을 인덱스로 하는 pandas.Series 자료형을 생성하시오.
단, 단위(%)는 포함하지 않습니다.

A  : 34.2 %
B  : 27.1 %
O  : 26.7 %
AB : 11.5 %

'''

### 정답

import pandas as pd

data = [34.2, 27.1, 26.7, 11.5]
ind = ['A', 'B', 'O', 'AB']

df = pd.Series(data=data, index=ind)
print(df)
```

```python
'''
# 문제4

아래의 단계별로 코드를 작성하시오.

# 1단계
np.arange() 함수를 사용하여 1 ~ 25 사이의 연속된 숫자(25개)를 가지는 넘파이 1차원 배열을 생성하시오.

# 2단계
1차원 넘파이 배열을 (5, 5)의 모양을 가지는 넘파이 2차원 배열을 생성하시오.

# 3단계
2단계에서 생성한 넘파이 2차원 배열을 이용해서 dataframe을 생성하시오.
단, 컬럼의 이름은 'a', 'b', 'c', 'd', 'e'로 설정하시오

'''

### 정답

# 1단계
arr = np.arange(1,26)
print(arr)

print('-'*80)

# 2단계
arr2 = arr.reshape((5,5))
print(arr2)
print(arr2.shape)

print('-'*80)

# 3딘계
df = pd.DataFrame(arr2, columns=['a','b','c','d','e'])
print(df)
```
