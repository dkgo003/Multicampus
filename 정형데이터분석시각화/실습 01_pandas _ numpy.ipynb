```python
### 필요한 라이브러리 임폴트

import pandas as pd
```

```python
### 리스트 자료형 데이터를 이용해서 Series 자료형 생성

data = [9904312, 3448737, 2890451, 2466052]
population = pd.Series(data)
print(population)
```

```python
### 인덱스를 지정해서 Series 자료형 생성

data = [9904312, 3448737, 2890451, 2466052]
index = ['서울','부산','인천','대구']
population = pd.Series(data=data, index=index)

# 결과 확인하기
print(population)
```

```python
#### dict 자료형을 이용하여 Series 자료형 생성

data = {"서울": 9904312, "부산": 3448737, "인천": 2890451, "대구": 2466052}
population = pd.Series(data)

# 결과 확인하기
print(population)
```

```python
### Series 데이터 확인

# values 확인
print(population.values)

print('-'*80)

# index 확인
print(population.index)

print('-'*80)

# dtype 확인
print(population.dtype)
```

```python
### 인덱싱

'''
1. index : 숫자, 문자열 인덱스를 이용하여 인덱싱
2. fancy indexing : 동시에 여러 개의 데이터를 인덱싱
'''

# Series 자료형 생성 --> population 활용
print(population)

print('-'*80)

# index 1번으로 인덱싱
print(f'두번째 원소 인덱싱 = {population[1]}')
print(f'두번째 원소 인덱싱 = {population["부산"]}')

print('-'*80)

# index 3번으로 인덱싱
print(f'네번째 원소 인덱싱 = {population[3]}')
print(f'네번째 원소 인덱싱 = {population["대구"]}')
```

```python
### 한번에 여러개를 리스트 형식을 이용하여 인덱싱 --> "fancy indexing"

'''
변수명[[index3, index0, index2]] --> 사용자가 지정한 index의 순서대로 값을 추출 가능
'''

# 리스트의 형식 [0, 3, 1]
fancy = population[[0, 3, 1]]
print(f'팬시 인덱싱의 결과 : \n{fancy}')

print('-'*80)

# 문자열 인덱스 --> fancy 인덱싱 가능
fancy = population[['서울', '인천', '부산']]
print(f'팬시 인덱싱의 결과 : \n{fancy}')
```

```python
### 슬라이싱

'''
1. 숫자 인덱스 : [start:end] --> start : end -1
2. 문자열 인덱스 : [start:end] --> start : end (이유 : 문자열 -1 (x))
'''

# 숫자 인덱스 이용
sliced = population[1:3]
print(sliced)

print('-'*80)

# 문자 인덱스 이용
sliced = population["부산":"대구"]
print(sliced)
```

```python
### 비교 연산자

# 두 변수 설정
a = 10
b = 3

# 두 변수의 비교
print(a > b)
print(a < b)
print(a >= b)
print(a <= b)
print(a == b)
print(a != b)
```

```python
### 논리 연산자

# 두 변수 설정
a = 10
b = 3

# & 연산자 : True and True --> True
condition1 = (a > b) & (a != b)
print(condition1)

print('-'*80)

# | 연산자 : True | False --> True
condition2 = (a > b) | (a == b)
print(condition2)
```

```python
### boolean 인덱싱

# 데이터 확인
print(df)

print('-'*80)

# 비교 연산자 적용
condition = (df['2015'] >= 3000000)
print(condition)
print(type(df['2015']))
print(type(df['2015'] >= 3000000))

print('-'*80)

# 비교 연산의 결과 --> 인덱싱 적용 --> fancy 인덱싱
print(df['2015'][condition])

print('-'*80)

# 논리 연산자 적용 -> 조건에 만족하는 행 추출
condition2 = (df['2015'] >= 3000000) & (df['2010'] >= 4000000)
print(f'condition2 결과 확인 : \n{condition2}')
```

```python
### boolean 인덱싱 보충

# 데이터프레임 자료형 데이터
print(df)

print('-'*80)

# 데이터프레임 --> '열'인덱싱 --> 시리즈 자료형 생성
myseries = df['2015']
print(f'자료형 : {type(myseries)}') 
print('-'*80)
print(f'값 : \n{myseries}')

print('-'*80)

# 2015 컬럼의 값과 3000000 값의 비교 연산
vals = myseries.values
print(f'2015 컬럼의 값 : {vals}')

print('-'*80)

res = []

for val in vals:
    print(f'val = {val}')
    # 비교 연산
    mid = (val >= 3000000)
    print(f'mid = {mid}')
    res.append(mid)

print(f'result = {res}')

print('-'*80)

# series 자료형 --> 컬럼 1개 --> '행' 인덱싱 만 존재!
fires = myseries[0]
print(f'0번째 인덱스에 대한 인덱싱의 결과 = {myseries[0]}')

print('-'*80)

# fancy 인덱싱 --> df['col][[index1, index2,...]]
fancy = myseries[[0, 1]]
print(f'fancy 인덱싱에 대한 결과 : \n{fancy}')

print('-'*80)

condition = [0, 1]
fancy1 = myseries[condition]
print(f'fancy 인덱싱에 대한 결과 : \n{fancy1}')

print('-'*80)

fancy2 = myseries[res]
print(f'boolean 인덱싱에 대한 결과 : \n{fancy2}')
```

```python
### concat 데이터 생성

data1 = {'A' : ['A0', 'A1', 'A2', 'A3'],
         'B' : ['B0', 'B1', 'B2', 'B3'],
         'C' : ['C0', 'C1', 'C2', 'C3'],
         'D' : ['D0', 'D1', 'D2', 'D3']}

data2 = {'A' : ['A4', 'A5', 'A6', 'A7'],
         'B' : ['B4', 'B5', 'B6', 'B7'],
         'C' : ['C4', 'C5', 'C6', 'C7'],
         'D' : ['D4', 'D5', 'D6', 'D7']}

data3 = {'A' : ['A8', 'A9', 'A10', 'A11'],
         'B' : ['B8', 'B9', 'B10', 'B11'],
         'C' : ['C8', 'C9', 'C10', 'C11'],
         'D' : ['D8', 'D9', 'D10', 'D11']}

df1 = pd.DataFrame(data=data1, index=[0,1,2,3])
df2 = pd.DataFrame(data=data2, index=[4,5,6,7])
df3 = pd.DataFrame(data=data3, index=[8,9,10,11])

# 결과 확인하기
print(df1)
print('-'*80)
print(df2)
print('-'*80)
print(df3)
```

```python
### concat() 함수 --> axis=0 방향으로 병합하기

concat1 = pd.concat([df1,df2,df3], axis=0)

# 결과 확인하기
print(concat1)

print('-'*80)

# concat --> 매개변수 : axis --> 기본값 = 0
concat2 = pd.concat([df1,df2,df3])
# 결과 확인하기
print(concat2)
```

```python
### concat 함수 --> axis=1 방향으로 병합
concat3 = pd.concat([df1,df2,df3], axis=1)

# 결과 확인하기
print(concat3)
```

## Numpy

```python
### 필요한 라이브러리 임폴트
import numpy as np
```

```python
### python list와 numpy nadrray 차이 --> 연산 가능 여부

# list 자료형 데이터 생성
list1 = [1,2,3]
list2 = [4,5,6]

# 리스트 자료형 --> 덧셈 연산 --> 불가능
result = list1 + list2
print(result)

print('-'*80)

# 리스트 자료형 --> 뺄셈 연산 --> 에러 발생
print(list1 - list2)
```

```python
## np.array() + 리스트 자료형 --> numpy 1차원 배열(1-d array)
arr1 = np.array(list1)
arr2 = np.array(list2)

# 결과 확인하기
print(f'배열 1 : {arr1}')

print('-'*80)

print(f'배열 2 : {arr2}')

print('-'*80)

# 덧셈 연산
res1 = arr1 + arr2
print(res1)

print('-'*80)

# 뺄셈 연산
res2 = arr1 - arr2
print(res2)
```

```python
### np.arange() 함수 이용 --> 1차원 배열(array) 만들기

# 0부터 10미만까지 1만큼 간격으로 정수의 1차원 배열 만들기
arr1 = np.arange(0, 10, 1)
print(arr1)
arr2 = np.arange(10) # 같다
print(arr2)

print('-'*80)
print(f'자료형 확인 : {type(arr1)}')
print('-'*80)

# 1부터 15 미만까지 2만큼의 간격으로 연속적인 정수의 1차원 배열 만들기
arr3 = np.arange(1, 15, 2)
print(arr3)
```

```python
### 2차원 배열(2-D array) 만들기

# 2차원 리스트 자료형 데이터 생성
mylist = [[1,2,3],[4,5,6],[7,8,9]]

# ndarray 생성 함수에 입력
arr = np.array(mylist)
print(arr)
```

```python
### 2차원 배열 만들기(2) --> np.arange().reshape(newshape=(m행,n열))

# np.arange() 사용 --> 1차원 배열 생성
arr1 = np.arange(12)
print(arr1)

print('-'*80)

# reshape() 사용 --> 1차원 배열을 (3, 4)의 2차원 배열로 변환
arr2 = arr1.reshape(3, 4)
print(arr2)

print('-'*80)

# 연속적인 콤보 적용
arr3 = np.arange(12).reshape(3, 4)
print(arr3)
```

```python
### 배열의 모양 확인 --> 변수.shape

# 1차원 배열 생성하기 : [1,2,.....10] --> 원소 10개의 1차원 배열
arr1 = np.arange(1,11)
print(arr1)

print('-'*80)

# 1차원 배열의 모양 확인 --> .shape
shape1 = arr1.shape
print(f'1차원 배열의 모양 : {shape1}')

print('-'*80)

# 2차원 배열 생성하기 --> arr1 재활용
arr2 = arr1.reshape(2,5)
print(arr2)

print('-'*80)

# 2차원 배열의 모양 확인 --> .shape
print(f'2차원 배열의 모양 : {arr2.shape}')
```

```python
### 보충 : pd.DataFrame() --> DataFrame 자료형 데이터 생성 --> df.shape

mydict = {'col1':[1,2],
          'col2':[3,4],
          'col3':[5,6]}

# 데이터프레임 자료형 데이터 생성
df = pd.DataFrame(data=mydict)
print(df)

print('-'*80)

# df.shape --> 모양 확인
print(f'생성된 데이터프레임의 모양 : {df.shape}')
```

```python
### 배열의 전체 원소 갯수 확인 --> 변수형.size

# 1차원 배열 생성
arr1 = np.arange(15)
print(arr1)

print('-'*80)

# 성분 원소의 갯수 확인(1)
print(f'배열의 성분 원소의 갯수 = {len(arr1)}')

print('-'*80)

# 성분 원소의 갯수 확인(2) --> size
print(f'배열의 성분 원소의 갯수 = {arr1.size}')

print('-'*80)

# 2차원 배열 생성
arr2 = np.arange(2,14).reshape(3,4)
print(arr2)

print('-'*80)

# 2차원 배열 --> 성분 원소의 갯수 확인 --> len() /.size
print(f'2차원 배열의 성분 원소의 갯수 = {len(arr2)}')
print('-'*80)
print(f'2차원 배열의 성분 원소의 갯수 = {arr2.size}')
```

```python
### 연습문제 3

mylist = [['3R', 1510, 7.36],
          ['3SOFT', 1790, 1.65],
          ['ACTS', 1185, 1.28]]

myindex = ['037730', '036360', '005670']

mycolumns = ['종목명', '현재가', '등락률']

df = pd.DataFrame(data=mylist, index=myindex, columns=mycolumns)

print(df)
```

```python
### 연습문제 4

# 4-1 --> loc 인덱서 사용
sliced1 = df.loc['037730':'036360', :]
print(sliced1)

print('-'*80)

# 4-1 --> iloc 인덱서 사용
sliced2 = df.iloc[:2, :]
print(sliced2)

print('-'*80)

# 4-1 --> fancy 인덱싱 응용
sliced3 = df.loc[['037730', '036360'], :]
print(sliced3)

print('-'*80)

# 4-2 --> loc 인덱서 사용
res = df.loc['037730', '현재가']
print(res)

print('-'*80)

# 4-2 --> iloc 인덱서 사용
res = df.iloc[0, 1]
print(res)
```

```python
### 배열의 차원 확인

import numpy as np

# 1차원 배열 생성
arr1 = np.arange(1,31,1)
print(arr1)

print('-'*80)

# 2차원 배열 생성 --> 1차원 배열 재구성
arr2 = arr1.reshape((3,10))
print(arr2)

print('-'*80)

# 1차원 배열의 모양과 차원 확인
print(f'1차원 배열의 모양 = {arr1.shape}')
print(f'1차원 배열의 차원 = {arr1.ndim}')

print('-'*80)

# 2차원 배열의 모양과 차원 확인
print(f'2차원 배열의 모양 = {arr2.shape}')
print(f'2차원 배열의 확인 = {arr2.ndim}')
```

```python
### 3차원 배열 생성 --> np.arange().reshape((행, 열, 깊이))

# 1차원 배열 생성
arr1 = np.arange(0,24,1)
print(arr1)

print('-'*80)

# reshape() 함수를 이용 --> (4, 3, 2)
arr2 = arr1.reshape((4, 3, 2))
print(arr2)
```

```python
### 연습문제

# 목표 : (2,2,2) 모양의 3차원 배열 생성

# 1차원 배열 생성
arr1 = np.arange(1,9)
print(arr1)

print('-'*80)

# 3차원 배열 생성
arr2 = arr1.reshape((2,2,2))
print(arr2)

print('-'*80)

# 모양 확인하는 코드
shape = arr2.shape
print(f'생성된 배열의 모양 = {shape}')

print('-'*80)

# 차원 확인하는 코드
n_dim = arr2.ndim
print(f'생성된 배열의 차원 = {n_dim}')

print('-'*80)

# 성분 원소의 갯수 확인하는 코드
n_size = arr2.size
print(f'생성된 배열의 원소의 수 = {n_size}')
```

```python
### 인덱싱 --> 특정 원소 접근

# 2차원 배열 생성하기
arr = np.arange(1,7).reshape((2,3))
print(arr)

print('-'*80)

# 행 인덱싱 --> 0번째 행 인덱싱 --> 0번째 행에 속하는 열은 모두 포함
row1 = arr[0]
print(f'행 인덱싱의 결과 = {row1}')

print('-'*80)

# 행렬 인덱싱 --> 0번째 행, 0번째 열 인덱싱
row_col1 = arr[0][0]
row_col2 = arr[0,0]
print(f'행렬 인덱싱의 결과(1) : {row_col1}')
print(f'행렬 인덱싱의 결과(2) : {row_col2}')
```

```python
### 3차원 배열에 대한 인덱싱

# 3차원 배열 생성 --> 모양(2,3,2)
arr = np.arange(12).reshape((2,3,2))
print(arr)

print('-'*80)

# 행 인덱싱 --> 1번행 인덱싱
row1 = arr[1]
row2 = arr[1, :]
print(f'1번행 인덱싱의 결과 : \n{row1}')
print(f'1번행 인덱싱의 결과 : \n{row2}')

print('-'*80)

# 행렬 인덱싱 --> 1번행 1번열 인덱싱
row_col1 = arr[1][1]
row_col2 = arr[1,1]
print(f'1번행 1번열 인덱싱의 결과 : {row_col1}')
print(f'1번행 1번열 인덱싱의 결과 : {row_col2}')

print('-'*80)

# 행렬깊이 인덱싱 --> 1번행, 1번열, 1번 깊이 인덱싱
row_col_depth1 = arr[1][1][1]
row_col_depth2 = arr[1,1,1]
print(f'1번행 1번열 1번 깊이 인덱싱의 결과 = {row_col_depth1}')
print(f'1번행 1번열 1번 깊이 인덱싱의 결과 = {row_col_depth2}')
```

```python
### 1차원 배열 슬라이싱

# np.arange() --> 0부터 9까지 10개의 정수로 구성된 1차원 배열 생성
arr = np.arange(10)
print(arr)

print('-'*80)

# 4번째 ~ 8번째 원소까지 가져오기
sliced = arr[3:8]
print(sliced)
```

```python
### 2차원 배열 슬라이싱

# 0부터 49까지 연속적인 정수의 1차원 배열 생성
arr1 = np.arange(50)
print(arr1)

print('-'*80)

# (5, 10) 모양의 2차원 배열 생성
arr2 = arr1.reshape((5,10))
print(arr2)

print('-'*80)

# 행 : 첫번째 두번째까지, 열 : 모든 열
sliced1 = arr2[0:2, :]
print(sliced1)

print('-'*80)

# 행 : 모든 행, 열 : 0번 열
sliced2 = arr2[:, 0]
print(sliced2)
```
