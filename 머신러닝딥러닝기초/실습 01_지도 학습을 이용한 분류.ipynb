### 데이터 생성하기

```python
### 필요한 라이브러리 임폴트
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
### 데이터 가져오기
iris_dataset = load_iris()

# 데이터 타입 확인하기
print(type(iris_dataset))
```

```python
### 데이터의 key, value 확인

'''
### sklearn.utils._bunch.Bunch 자료형의 특징 : 사전(dict)처럼 keys(), values(), items() 제공
'''

# 데이터의 keys(목차) 확인
print(iris_dataset.keys())

print('-'*80)

# 데이터의 values 확인
print(iris_dataset.values())
```

```python
### 분류에 사용할 특성(feature, column)의 종류와 이름 확인 --> 학습에 사용할 데이터(X)의 종류와 이름
print(f'특성의 종류 = {len(iris_dataset.feature_names)}')
print('-'*80)
print(f'특성의 이름 = {iris_dataset.feature_names}')
```

```python
### 학습에 사용할 데이터 4가지의 값(values) 확인
print(f'학습에 사용할 데이터 : \n{iris_dataset.data}')
print('-'*80)
print(f'학습에 사용할 데이터의 자료형 : {type(iris_dataset.data)}')
print('-'*80)
print(f'학습에 사용할 데이터의 모양 : \n{iris_dataset.data.shape}')
```

```python
### 목표 --> 붓꽃의 품종 분류 --> 품좀의 이름
print(f'품종의 이름 = {iris_dataset.target_names}')
```

```python
### 붓꽃의 품종의 값과 자료형 확인 --> 모델 성능 평가 --> 정답
print(f'분류의 결과 값 : \n{iris_dataset.target}')
print('-'*80)
print(f'분류의 결과 값의 자료형 = {type(iris_dataset.target)}')
print('-'*80)
print(f'분류의 결과 값의 모양 = {iris_dataset.target.shape}')
```

```python
### 데이터 프레임 생성

# data와 column 이름 설정
iris_data = iris_dataset.data
iris_columns = iris_dataset.feature_names

# 데이터프레임 생성
df = pd.DataFrame(data=iris_data, columns=iris_columns)

# 결과 확인하기
df
```

```python
### label 컬럼 생성 --> 붓꽃 품종의 값 삽입
df['label'] = iris_dataset.target

# 결과 확인하기
df
```

```python
### 붓꽃 품종의 값 --> 이름으로 대체 --> replace() 사용
df['label'] = df['label'].replace({0:iris_dataset.target_names[0], 1:iris_dataset.target_names[1], 2:iris_dataset.target_names[2]})

# 결과 확인하기 
print(f'데이터의 모양 = {df.shape}')
print('-'*80)
print(f'전체 데이터 확인 : \n{df}')
print('-'*80)
print('데이터의 기본 정보')
df.info()
```

```python
### 붓꽃 품종의 종류별 개수 확인 --> '정답'의 분포 확인

# 정답의 분포 추출
counts = df['label'].value_counts()
print(counts)

print('-'*80)

# 분포 시각화
sns.countplot(data=df, x='label')
plt.show()
```

```python
### 각 품종별 특성 확인 --> boxplot() 사용 --> 시각화

## loc 인덱서 + boolean 배열 --> 품종별로 데이터프레임 분할

# setosa
df_setosa = df.loc[df['label']=='setosa', :]
# versicolor
df_versicolor = df.loc[df['label']=='versicolor', :]
# virginica
df_virginica = df.loc[df['label']=='virginica', :]

## boxplot() 사용
# setosa
print('setosa의 특성')
df_setosa.plot(kind='box', figsize=(10,6))
plt.show()

print('-'*80)

# versicolor
print('versicolor의 특성')
df_versicolor.plot(kind='box', figsize=(10,6))
plt.show()

print('-'*80)

# virginica
print('virginica의 특성')
df_virginica.plot(kind='box', figsize=(10,6))
plt.show()
```

```python
### 각 특성별 품종 간의 차이 확인 --> sns.kdeplot() 사용 --> 시각화

'''
1. 범주형 변수 값 --> 상대적 비율 : value_counts(normalize=True) --> pie chart, bar chart
2. 연속형 변수 값 --> 상대적 비율 --> 연속적인 그래프 --> sns.kdeplot(data, x, hue)
'''

# 품종별 꽃받침의 길이 비교
print('품종별 꽃받침의 길이 비교')
sns.kdeplot(data=df, x=df.columns[0], hue='label')
plt.show()

print('-'*80)

# 품종별 꽃받침의 너비 비교
print('품종별 꽃받침의 너비 비교')
sns.kdeplot(data=df, x=df.columns[1], hue='label')
plt.show()

print('-'*80)

# 품종별 꽃잎의 길이 비교
print('품종별 꽃잎의 길이 비교')
sns.kdeplot(data=df, x=df.columns[2], hue='label')
plt.show()

print('-'*80)

# 품종별 꽃잎의 너비 비교
print('품종별 꽃잎의 너비 비교')
sns.kdeplot(data=df, x=df.columns[3], hue='label')
plt.show()
```

### 학습용 데이터와 평가용 데이터 생성

```python
### 75:25의 비율로 학습용 데이터와 평가용 데이터 생성

# 필요한 라이브러리 임폴트
from sklearn.model_selection import train_test_split

# X 데이터 설정
X = iris_dataset.data

# y 데이터 설정
y = iris_dataset.target

# train_test_split() 사용
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=13)
```

```python
### 학습용 데이터의 자료형과 형태 확인
print(f'학습용 데이터의 자료형 = {type(X_train)}')
print('-'*80)
print(f'학습용 데이터의 모양 = {X_train.shape}')
```

```python
### 평가용 데이터의 자료형과 형태 확인
print(f'평가용 데이터의 자료형 = {type(X_test)}')
print('-'*80)
print(f'평가용 데이터의 모양 = {X_test.shape}')
```

### Decision Tree 모델을 이용한 분류

#### 모델 생성 및 학습

```python
### Decision Tree 분류 모델 생성

# 필요한 라이브러리 임폴트
from sklearn.tree import DecisionTreeClassifier

# 생성 함수 호출, 모델 객체 생성
dt_iris = DecisionTreeClassifier(random_state=13)
```

```python
### Decision Tree 분류 모델 학습
dt_iris.fit(X_train, y_train)
```

```python
### 학습 --> 조건 확인
dt_iris.get_params()
```

#### 학습용 데이터와 평가용 데이터를 이용한 예측

```python
### 임의의 데이터를 이용한 품종 분류 예측

'''
꽃받침의 길이가 5cm, 폭이 2.9cm 이고 꽃잎의 길이가 1cm, 폭이 0.2cm 인 붓꽃 데이터 생성
데이터는 반드시 연산이 가능한 'numpy.ndarray'이어야 한다
데이터 입력시 예상되는 품종은?
'''

# 입력 데이터 생성
X_new = np.array([[5, 2.9, 1, 0.2]])

# 학습된 모델을 이용하여 예측
prediction = dt_iris.predict(X_new)

# 결과 확인하기
print(f'예측 결과값 = {prediction}')
print('-'*80)
print(f'예상되는 품종은? : {iris_dataset.target_names[prediction[0]]}')
```

```python
### 학습용 데이터에 대해서 예측하기
pred_train = dt_iris.predict(X_train)

# 결과 확인하기
print(pred_train)
```

```python
### 평가용 데이터에 대해서 예측하기
pred_test = dt_iris.predict(X_test)

# 결과 확인하기
print(pred_test)
```

#### 정확도 평가하기

```python
# 정확도를 평가하기 위해서 필요한 라이브러리 임폴트
from sklearn.metrics import accuracy_score

# 학습용 데이터에 대해서 정확도 평가
accuracy_train = accuracy_score(y_train, pred_train)

# 평가용 데이터에 대해서 정확도 평가
accuracy_test = accuracy_score(y_test, pred_test)

# 결과 확인하기
print(f'학습용 데이터에 대한 정확도 = {accuracy_train}')

print('-'*80)
print(f'평가용 데이터에 대한 정확도 = {accuracy_test}')
```

#### GridSearchCV를 이용한 모델의 성능 개선 (4/17 실습)

```python
### 필요한 라이브러리 임폴트
from sklearn.model_selection import GridSearchCV
```

```python
### GridSearchCV 함수 실행

# DecisionTree 분류 기본 모델 생성
base_model = DecisionTreeClassifier(random_state=13)

# 튜닝할 매개 변수 설정
myparams = np.arange(3,8)
params = {'max_depth':myparams}

# GridSearchCV 클래스 호출, 객체 생성
grid_iris = GridSearchCV(
    estimator=base_model,
    param_grid=params,
    cv=10,
    scoring='accuracy'
)

# GridSearchCV 객체 --> 학습 및 평가 수행
grid_iris.fit(X_train, y_train)
```

```python
### 가장 높은 성능 --> 하이퍼파라미터 확인
print(grid_iris.best_params_)
```

```python
### best 모델 완성
best_model_iris = DecisionTreeClassifier(max_depth=3, random_state=13)
```

```python
### best 모델 학습
best_model_iris.fit(X_train, y_train)
```

```python
### 학습용 데이터 / 평가용 데이터 --> 예측

# 학습용 데이터에 대해서 예측
pred_train = best_model_iris.predict(X_train)

# 평가용 데이터에 대해서 예측
pred_test = best_model_iris.predict(X_test)
```

```python
### 정확도 평가

# 정확도 평가를 위해서 필요한 라이브러리 임폴트
from sklearn.metrics import accuracy_score

# 학습용 데이터에 대한 정확도 평가
accuracy_train = accuracy_score(y_train, pred_train)

# 평가용 데이터에 대한 정확도 평가
accuracy_test = accuracy_score(y_test, pred_test)

# 결과 확인하기
print(f'학습용 데이터에 대한 정확도 = {accuracy_train}')
print('-'*80)
print(f'평가용 데이터에 대한 정확도 = {accuracy_test}')
```
