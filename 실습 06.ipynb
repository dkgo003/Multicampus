## 세트(set)

```python
# 세트 자료형 데이터 생성
test_set = {1, 2, 3, 4, 5, 1, 2}
print(test_set)

# 세트 자료형 데이터 생성2
test_set2 = set([1, 2, 3, 1, 2, 3])
print(test_set2)
```

```python
# 여러 자료형 생성 함수 정리
test = 'abcdabc'
print(test)

print('-'*80)

test_list = list(test)
print(test_list)

print('-'*80)

test_tuple = tuple(test)
print(test_tuple)

print('-'*80)

test_set = set(test)
print(test_set)

print('-'*80)

print(list(test_tuple))
```

```python
### QUIZ

# 데이터가 없는 리스트 생성 --> 가능
result = []
result.append(7)
print(result)

# 데이터가 없는 튜플 생성 --> 가능
result1 = tuple()
print(result1)

# 데이터가 없는 집합 생성 --> 가능
result2 = {}
print(result2)
```

```python
### 세트는 튜플과 다르게 삭제나 변경이 가능함

s = set([1, 2, 3, 1, 2, 3])

# 세트는 튜플과 다르게 삭제나 변경이 가능함
s.add(1)
print(s)
s.remove(1)
print(s)
s.update([1, 4, 5, 6, 7])
print(s)
s.discard(3)
print(s)
s.clear()
print(s)
```

```python
### 파이썬은 교집합, 합집합, 차집합 연산을 모두 지원함

s1 = set([1, 2, 3, 4, 5])
s2 = set([3, 4, 5, 6 ,7])

# 합집합
print(s1.union(s2)) # s1과 s2의 합집합
print(s1 | s2) # set([1, 2, 3, 4, 5, 6, 7])

# 교집합
print(s1.intersection(s2)) # s1과 s2의 교집합
print(s1 & s2) # set([3, 4, 5])

# 차집합
print(s1.difference(s2)) # s1과 s2의 차집합
print(s1 - s2) # set([1, 2])
```

## 딕셔너리(dict)

```python
### 국가명과 국가 전화번호를 묶어 보여주는 코드 작성

country_code = { } # 딕셔너리 생성
country_code = {"America": 1, "Korea": 82, "China": 86, "Japan": 81}
print(country_code)
```

```python
### key=value(변수 = data 형식) 형식으로 딕셔너리 생성

'''
1. dict() 생성 함수 이용
2. key가 문자열이라 하더라도 key에 ''나 ""사용하지 않아야 한다
'''

country_code = dict(America=1, Korea=82, China=86, Japan=81)
print(country_code)
```

```python
### 딕셔너리 변수 안의 키와 값을 출력하는 함수

keys = country_code.keys()
print(keys)

# contry_code_keys()의 특성 --> 인덱싱 가능 여부 --> 불가능
# print(keys[0])

# contry_code_keys()의 특성 --> for문에서 사용 가능 여부 --> 가능
for key in keys:
    print(key)
```

```python
### 딕셔너리 변수 안의 값을 출력하는 함수

values = country_code.values()
print(values)

# contry_code_values()의 특성 --> 인덱싱 가능 여부 --> 불가능
# print(values[0])

# contry_code_values()의 특성 --> for문에서 사용 가능 여부 --> 가능
for value in values:
    print(value)
```

```python
### 딕셔너리 변수 안의 키-값을 동시에 출력하는 함수

items = country_code.items()
print(items)

# contry_code_items()의 특성 --> 인덱싱 가능 여부 --> 불가능
# print(items[0])

# contry_code_items()의 특성 --> for문에서 사용 가능 여부 --> 가능
for k,v in items:
    print(f'key = {k}')
    print(f'value = {v}')
```

```python
### 딕셔너리 추가

country_code['German'] = 49

# 결과 확인하기
print(country_code)
```

```python
### 딕셔너리 수정

country_code['German'] = 1

# 결과 확인하기
print(country_code)
```

```python
### 특정 키나 값이 해당 변수에 포함되어 있는지 확인하기

# 특정 키의 존재 여부 확인
print("Korea" in country_code.keys())

# 특정 값의 존재 여부 확인
print(82 in country_code.values())
```

## collections 모듈

### OrderedDict 모듈

```python
### OrderedDict와 python dict 비교

'''
1. collections 모듈에서 제공하는 OrderedDict() : 순서가 있는 딕셔너리
2. python dict : 3.6버전부터 순서를 부여
'''

from collections import OrderedDict

# python dict 자료형 데이터 생성
python_dict = {}
# print(python_dict)
python_dict['x'] = 100
python_dict['y'] = 200
python_dict['z'] = 300
python_dict['l'] = 500

for k, v in python_dict.items():
    print(f'k : {k},  v : {v}')

print('-'*80)

ordered_dict = OrderedDict()
ordered_dict['x'] = 100
ordered_dict['y'] = 200
ordered_dict['z'] = 300
ordered_dict['i'] = 500

for k, v in ordered_dict.items():
    print(f'k : {k},  v : {v}')
```

### Counter 모듈

```python
### 시퀀스 자료형의 데이터 값의 개수를 딕셔너리 형태로 전달

from collections import Counter

# 리스트 자료형 데이터 생성
text = list('gallahad')
print(text)

print('-'*80)

# Counter 자료형 데이터 생성
count = Counter(text)
print(count)

print('-'*80)

# 키로 검색 가능
print(count['a'])
```

```python
### split() 함수의 기능

'''
매개 변수가 없으면 공백을 기준으로 문자열을 분리한다
'''

text = """A press release is the quickest and easiest way to get free publicity. If well written, a press release can result in multiple published articles about your firm and its products. And that can mean new prospects contacting you asking you to sell to them. Talk about low-hanging fruit!
What's more, press releases are cost effective. If the release results in an article that (for instance) appears to recommend your firm or your product, that article is more likely to drive prospects to contact you than a comparable paid advertisement.
However, most press releases never accomplish that. Most press releases are just spray and pray. Nobody reads them, least of all the reporters and editors for whom they're intended. Worst case, a badly-written press release simply makes your firm look clueless and stupid.
For example, a while back I received a press release containing the following sentence: "Release 6.0 doubles the level of functionality available, providing organizations of all sizes with a fast-to-deploy, highly robust, and easy-to-use solution to better acquire, retain, and serve customers."
Translation: "The new release does more stuff." Why the extra verbiage? As I explained in the post "Why Marketers Speak Biz Blab", the BS words are simply a way to try to make something unimportant seem important. And, let's face it, a 6.0 release of a product probably isn't all that important.
As a reporter, my immediate response to that press release was that it's not important because it expended an entire sentence saying absolutely nothing. And I assumed (probably rightly) that the company's marketing team was a bunch of idiots."""

print(text)

splited_list = text.lower().split()

print(splited_list)
```

```python
### Counter 자료형을 이용하여 각 단어의 개수 세는 작업 실행

word_count = Counter(splited_list)

# 결과 확인하기
print(word_count)
```
