```python
### [코드 5-1]

# 사각형의 넓이를 구하는 함수 정의
def calculate_rectangle_area(x, y):
    return x * y

rectangle_x = 10
rectangle_y = 20
print(f"사각형 x의 길이 : {rectangle_x}")
print(f"사각형 y의 길이 : {rectangle_y}")

# 사각형의 넓이를 구하는 함수 호출
area = calculate_rectangle_area(rectangle_x, rectangle_y)
print(f"사각형의 넓이 : {area}")
```

```python
### [코드 5-2]

# 함수 f(x), g(x)에 대한 정의
def f(x):
    return 2 * x + 7
def g(x):
    return x ** 2

# 매개변수(parameter) x에 전달하는 값 --> 인수(인자, argument)
x = 2

# 함수 실행
res = f(x) + g(x) + f(g(x)) + g(f(x))
print(res)
```

```python
### [코드 5-8] ==> 지역 변수와 전역 변수는 서로가 서로에게 아무런 영향을 주지 못한다

def f():
    # 지역변수 s
    s = "I love London!"
    print(s)

# 전역변수 s
s = "I love Paris!"
f()
print(s)
```

```python
### [코드 5-9]

def f():
    global s  # 지역 변수를 전역 변수로 사용하게 해줌
    s = "I love London!"
    print(s)

s = "I love Paris!"
print(s)
f()
print(s)
```

```python
### [코드 5-10]

def calculate(x, y):
    total = x + y              # 새로운 값이 할당되어 함수 내부 total은 지역 변수가 됨
    print("In Function")
    print(f"a : {str(a)}", f"b : {str(b)}", f"a + b : {str(a + b)}", f"total : {str(total)}")
    return total

a = 5                          # a와 b는 전역 변수
b = 7
total = 0                      # 전역 변수 total
print("In Program - 1")
print(f"a : {str(a)}", f"b : {str(b)}", f"a + b : {str(a + b)}")

sum = calculate(a, b)
print("After Calculation")
print(f"total : {str(total)}", f"Sum : {str(sum)}")
```

```python
### [코드 5-11]

def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n - 1)

res = factorial(int(input("Input Number for Factorial Calculation : ")))
print(res)
```

```python
### 과제

### 평균 구하기

kor_score = [49, 80, 20, 100, 80]
math_score = [43, 60, 85, 30, 90]
eng_score = [49, 82, 48, 50, 100]
midterm_score = [kor_score, math_score, eng_score]

student_score = [0, 0, 0, 0, 0]
i = 0
for subject in midterm_score:
    print('-' * 40)
    print(f'첫번째 for문 : i = {i}')
    print('-' * 40)
    for score in subject:
        print(f'두번째 for문 : i = {i}')
        student_score[i] += score   # 학생마다 개별로 교과 점수를 저장
        i += 1  # 학생 인덱스 구분
    i = 0   # 과목이 바뀔 때 학생 인덱스 초기화
else:
    a, b, c, d, e = student_score   # 학생별 점수를 언패킹
    student_score = [a/3, b/3, c/3, d/3, e/3]
    print(student_score)
```
