## 1. 보스턴 집값 예측

### 4/20 실습

### 데이터 생성하기

```python
### 필요한 라이브러리 임폴트
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
### 데이터 가져오기
file_path = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv'
df_boston = pd.read_csv(file_path)

# 결과 확인하기
df_boston
```

### 학습용 데이터와 평가용 데이터 생성

```python
### 75:25의 비율로 학습용 데이터와 평가용 데이터 생성

# 필요한 라이브러리 임폴트
from sklearn.model_selection import train_test_split

# X 데이터 생성
X_data = df_boston.drop(columns=['medv'])

# y 데이터 생성 medv --> median value of owner-occupied homes
y_data = df_boston.medv

# 학습용 / 평가용 데이터 생성
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.25, random_state=20)
```

```python
### 학습용 데이터의 자료형과 형태 확인
print(f'학습용 X 데이터의 자료형 : {type(X_train)}')
print(f'학습용 X 데이터의 모양 : {X_train.shape}')
print('-'*80)
print(f'학습용 y 데이터의 자료형 : {type(y_train)}')
print(f'학습용 y 데이터의 모양 : {y_train.shape}')
```

```python
### 평가용 데이터의 자료형과 형태 확인
print(f'평가용 X 데이터의 자료형 : {type(X_test)}')
print(f'평가용 X 데이터의 모양 : {X_test.shape}')
print('-'*80)
print(f'평가용 y 데이터의 자료형 : {type(y_test)}')
print(f'평가용 y 데이터의 모양 : {y_test.shape}')
```

### 데이터 스케일링 적용

```python
# 필요한 라이브러리 임폴트
from sklearn.preprocessing import MinMaxScaler

# MinMaxScaler 객체 생성
scaler = MinMaxScaler()

# scaling 적용(1) --> 학습용 데이터 --> mix / max 파악
scaler.fit(X_train)

# scaling 적용(2) --> 데이터 변형(scaling)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
### 변형된 결과 확인하기

# 변환된 데이터의 자료형 확인
print(f'변형된 데이터의 자료형 = {type(X_train_scaled)}')

print('-'*80)

# 최소 / 최대 값 확인
print(f'변형된 학습용 데이터의 최소값 : {np.min(X_train_scaled)}')
print('-'*80)
print(f'변형된 학습용 데이터의 최대값 : {np.max(X_train_scaled)}')
```

### LinearRegression 모델

#### 모델 생성 및 학습

```python
### 모델 생성

# 필요한 라이브러리 임폴트
from sklearn.linear_model import LinearRegression

# 모델 객체 생성
lr = LinearRegression()
lr_scaled = LinearRegression()
```

```python
### 모델 학습(1) --> 스케일링이 되지 않은 데이터
lr.fit(X_train, y_train)
```

```python
### 모델 학습(2) --> 스케일링이 된 데이터
lr_scaled.fit(X_train_scaled, y_train)
```

```python
### 학습의 결과 생성된 선형 모델의 기울기 / 절편 확인(1) --> 스케일링이 되지 않은 데이터
print(f'선형 모델의 직선의 기울기 : \n{lr.coef_}')
print('-'*80)
print(f'선형 모델의 직선의 절편 : \n{lr.intercept_}')
```

```python
### 학습의 결과 생성된 선형 모델의 기울기 / 절편 확인(2) --> 스케일링이 된 데이터
print(f'선형 모델의 직선의 기울기 : \n{lr_scaled.coef_}')
print('-'*80)
print(f'선형 모델의 직선의 절편 : \n{lr_scaled.intercept_}')
```

#### 학습용 / 평가용 데이터를 이용한 예측

```python
### 데이터 : 스케일링이 되지 않은 데이터

# 학습용 데이터를 이용한 예측
pred_train = lr.predict(X_train)

# 평가용 데이터를 이용한 예측
pred_test = lr.predict(X_test)
```

```python
### 데이터 : 스케일링이 된 데이터

# 학습용 데이터를 이용한 예측
pred_train_scaled = lr_scaled.predict(X_train_scaled)

# 평가용 데이터를 이용한 예측
pred_test_scaled = lr_scaled.predict(X_test_scaled)
```

#### 성능 평가

```python
### 데이터 : 스케일링이 되지 않은 데이터

# 모델의 성능을 평가하기 위한 라이브러리 임폴트
from sklearn.metrics import mean_squared_error

# 학습용 데이터에 대한 평가
mse_train = mean_squared_error(y_train, pred_train)

# 평가용 데이터에 대한 평가
mse_test = mean_squared_error(y_test, pred_test)

# 결과 확인하기
print(f'학습용 데이터에 대한 성능 평가 = {mse_train}')
print('-'*80)
print(f'평가용 데이터에 대한 성능 평가 = {mse_test}')
```

```python
### 데이터 : 스케일링이 된 데이터

# 모델의 성능을 평가하기 위한 라이브러리 임폴트
from sklearn.metrics import mean_squared_error

# 학습용 데이터에 대한 평가
mse_train_scaled = mean_squared_error(y_train, pred_train_scaled)

# 평가용 데이터에 대한 평가
mse_test_scaled = mean_squared_error(y_test, pred_test_scaled)

# 결과 확인하기
print(f'스케일링이 된 학습용 데이터에 대한 성능 평가 = {mse_train_scaled}')
print('-'*80)
print(f'스케일링이 된 평가용 데이터에 대한 성능 평가 = {mse_test_scaled}')
```
