```python
### 문자열의 슬라이싱

# 문자열 데이터 생성
a = "TEAMLAB MOOC, AWESOME Python"

# a 변수의 0부터 5까지, -9부터 끝까지
print(a[0:6], " AND ", a[-9:])

print('-'*80)

# a 변수의 처음부터 끝까지
print(a[:])

print('-'*80)

# 범위를 넘어갈 경우 자동으로 최대 범위를 지정
print(a[-50:50])

print('-'*80)

# 2칸씩, 역순으로
print(a[::2], " AND ", a[::-1])
```

```python
### 문자열의 연산(1)

a = "TEAM"
b = "LAB"

# 덧셈으로 a와 b 변수 연결하기
print(a + " " + b)

print('-'*80)
      
# 곱하기로 반복 연산 가능
print(a * 2 + " " + b * 2)

print('-'*80)

# 'A'가 a에 포함되었는지 확인
if 'A' in a:
    print(a)
else:
    print(b)
```

```python
### 문자열 함수(1)

title1 = "TEAMLAB X Inflearn"

# title 변수를 모두 대문자로 변환
upper = title1.upper()
print(upper)

print('-'*80)

# title 변수를 모두 소문자로 변환
lower = title1.lower()
print(lower)

print('-'*80)

title2 = "TEAMLAB X Inflearn"

# title 변수의 각 단어의 앞글자만 대문자로 변환
print(title2.title())

print('-'*80)

#  title 변수의 첫 번째 글자만 대문자로 변환
print(title2.capitalize())
```

```python
### 문자열 함수(2)

title = "TEAMLAB X Inflearn"

# title 변수에 ‘a’의 개수 반환
print(title.count('a'))

print('-'*80)

# title 변수를 대문자로 만든 후, ‘a’의 개수 반환
print(title.upper().count('a'))

print('-'*80)

# title 변수의 문자열이 숫자인지 여부 반환
print(title.isdigit())

print('-'*80)

# title 변수가 ‘a’로 시작하는지 여부 반환
print(title.startswith('a'))
```

```python
### 파일 읽기(1)

# 파일 경로 설정
file_path = '/content/drive/MyDrive/KDT/python/yesterday.txt'

# 파일 열기
f = open(file_path, 'r')
print(f)

print('-'*80)

# 파일 읽기 --> readlines()
yesterday_lyric1 = f.readlines()

# 파일 닫기
f.close()

# 결과 출력
print(yesterday_lyric1)
print('-'*80)
print(f'readlines() 함수로 읽은 결과의 자료형 : {type(yesterday_lyric1)}')
print('-'*80)

# 결과 --> 문자열의 리스트 --> 인덱싱 또는 슬라이싱 가능
first_line = yesterday_lyric1[0]
print(f'첫번째 줄 : \n{first_line}')
print('-'*80)
print(f'첫번째 줄의 자료형 : {type(first_line)}')
```

```python
### 파일 읽기(2)

# 파일 열기
f = open(file_path, 'r')

# 파일 읽기 --> read
yesterday_lyric2 = f.read()

# 파일 닫기
f.close()

# 결과 출력
print(yesterday_lyric2)
print('-'*80)
print(f'read() 함수로 읽은 결과의 자료형 : {type(yesterday_lyric2)}')
```

```python
### strip() 함수

'''
strip() : 문자열의 시작과 끝에 있는 줄바꿈(\n)과 공백을 제거
'''

string = "\n Hello World. Python! \n"

new_string = string.strip()

print(f'줄바꿈과 공백 제거 전 : {string}')

print('-'*80)

print(f'제거 후 : {new_string}')
```

```python
### [코드 6-1]

contents = ""
for line in yesterday_lyric1:
    # print(f'line 확인 : {line}')
    contents = contents + line.strip() + "\n" 
    # readlines() 함수를 사용하면 필연적으로 1줄의 줄바꿈이 생기기때문에 strip() 함수로 공백을 자워주는 작업을 해야한다.
    # print(f'contests 확인 : {contents}')

n_of_yesterday = contents.upper().count("YESTERDAY")
print(f"Number of a Word 'Yesterday' : {n_of_yesterday}")
```
