## re(정규 표현식)
- 특정 규칙(패턴)이 있는 텍스트 데이터를 빠르게 처리할 수 있도록 도와주는 라이브러리

### 5/4 실습

```python
### 필요한 라이브러리 임폴트
import re
```

```python
### 패턴 정의 / 정리

# 문자 또는 숫자를 찾을 수 있는 패턴
pattern1 = '\w'

# 영어 알파벳을 찾을 수 있는 패턴
pattern2 = '[a-zA-Z]' 

# 영어 알파벳 또는 숫자를 찾을 수 있는 패턴
pattern3 = '[a-zA-Z0-9]'

# 한글(자음, 모음, 글자)을 찾을 수 있는 패턴 
pattern4 = '[ㄱ-ㅎㅏ-ㅣ가-힣 ]'

# 한글(글자)을 찾을 수 있는 패턴
pattern5 = '[가-힣]'

# 한글(글자) 또는 숫자를 찾을 수 있는 패턴
pattern6 = '[가-힣0-9]'

# 숫자를 찾을 수 있는 패턴(1)
pattern7 = '[0-9]'

# 숫자를 찾을 수 있는 패턴(2)
pattern8 = '\d' 

# 한글(글자)을 제외한 모든 문자 또는 숫자 또는 기호를 찾을 수 있는 패턴
pattern9 = '[^가-힣 ]'

# 한글(글자) 또는 숫자를 제외한 모든 문자 또는 기호를 찾을 수 있는 패턴
pattern10 = '[^가-힣0-9]'

# 영어 알파벳 또는 숫자를 제외한 모든 문자 또는 기호를 찾을 수 있는 패턴
pattern11 = '[^a-zA-Z0-9]'

# 숫자를 제외한 나머지 문자 또는 기호를 찾을 수 있는 패턴
pattern12 = '[^0-9]'  
```

```python
### 함수 사용법(1)

'''
search() : 문자열 전체에 대해서 특정 패턴과 매치되는 부분이 있는지 검색
'''
text1 = 'wow, it is awesome'
result1 = re.search(pattern1, text1)

# 모든 검색 결과 출력 --> findall()
# result = re.findall(pattern1, text)

text2 = '패딩(padding)'
result2 = re.search(pattern1, text2)

# 결과 확인하기
print(result1)
print('-'*80)
print(result2)
```

```python
### 함수 사용법(2)

'''
split() : 패턴을 기준으로 문자열을 분리하여 리스트 자료형 생성
'''
pattern = '\W' 
text = 'it is world of word'

# result = re.split(pattern, text)
result = text.split()

# 결과 확인하기
print(result)
```

```python
### 함수 사용법

'''
sub() : 문자열에서 패턴과 일치하는 부분에 대해서 다른 문자열로 대체
'''

result = re.sub(pattern8, 'number', '7 candy')

# 결과 확인하기
print(result)
```
