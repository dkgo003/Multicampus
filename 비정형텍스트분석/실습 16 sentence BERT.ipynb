```python
### 필요한 라이브러리 임폴트
from sentence_transformers import SentenceTransformer
```

### 한글 문장 유사도 측정

```python
### 문장 임베딩 벡터를 추출 --> 사전 학습된 한국어 SentenceTransformer(Sentence-BERT) 모델 생성
model_name='ddobokki/klue-roberta-base-nli-sts'
model_kor = SentenceTransformer(model_name) 
```

```python
### 한글 텍스트 데이터 생성
ko_sentences = ["직원이 무단 퇴사를 했는데 손해 배상 청구할 수 있나요?", "무단 퇴사한 직원에 대한 손해 배상 청구가 가능한가요?"]
```

```python
### 문장 임베딩 벡터 생성하기
embeddings = model_kor.encode(ko_sentences)

# 결과 확인하기
print(f'실습용 문장의 임베딩 벡터의 모양 : {embeddings.shape}')
print('-'*80)
print(f'실습용 문장 1의 임베딩 벡터 : \n{embeddings[0]}')
print('-'*80)
print(f'실습용 문장 2의 임베딩 벡터 : \n{embeddings[1]}')
```

```python
### 코싸인 유사도 측정

# 필요한 라이브러리 임폴트
from sklearn.metrics.pairwise import cosine_similarity

# 1차원 배열 --> 2차원 배열로 변환
embedding1 = embeddings[0].reshape((1,768))
embedding2 = embeddings[1].reshape((1,768))

# cosine_similarity() 사용
similarity = cosine_similarity(embedding1, embedding2)

# 결과 확인하기
print(similarity)
```

```python
### 필요한 라이브러리 임폴트
import pandas as pd
from sklearn.metrics.pairwise import linear_kernel
```

```python
### pd.read_csv() --> DataFrame 자료형 생성
file_path = 'movies_metadata.csv'
df = pd.read_csv(file_path)
print(df)
print('-'*80)
df.info()
```

```python
### 줄거리(overview), 제목(title) 컬럼을 기준 --> 누락 데이터 제거

# fancy indexing
data = df[['overview', 'title']]

# 누락 데이터 제거
df_movies = data.dropna()

# 결과 확인하기
print(df_movies)
print('-'*80)
df_movies.info()
```

```python
### 누락 데이터 제거 후 인덱스 재조정 --> reset_index(drop=True)
df_movies.reset_index(drop=True, inplace=True)

# 결과 확인하기
print(df_movies)
```

```python
### data의 수를 10000개로 축소
df_data = df_movies.iloc[:10000, :] 

# 결과 확인하기
print(df_data)
```

```python
### SentenceTransformer 모델 객체 생성
model_name = 'sentence-transformers/all-MiniLM-L12-v2'
model = SentenceTransformer(model_name)
```

```python
### 문장 임베딩 벡터 생성

# 필요한 라이브리 임폴트
import numpy as np

# overview 컬럼 --> 배열 생성
sentences = df_data.overview.values

# model.encode() 함수 사용
embeddings = model.encode(sentences)

# 결과 확인하기
print(f'생성된 문장 임베딩 행렬의 모양 : {embeddings.shape}')
print('-'*80)
print(f'문장 1개의 임베딩 벡터의 모양 : {embeddings[0].shape}')
```

```python
### 문장 임베딩 행렬 저장
np.save('embeddings.npy', embeddings)
```

```python
### linear_kernel() 함수 --> 10000개의 영화 줄거리 각각에 대해서 코사인 유사도 측정
similarity = linear_kernel(embeddings, embeddings)

# 결과 확인하기
print(f'유사도 측정의 결과 : \n{similarity}')
print('-'*80)
print(f'유사도 측정의 결과 값의 모댱 : {similarity.shape}')
print('-'*80)
print(f'Toy Story에 대한 유사도 측정의 결과 : \n{similarity[0]}')
```

```python
### 코사인 유사도 --> 데이터프레임 생성
columns = df_data.title.values
index = df_data.title.values

# 데이터프레임 생성
df_sim = pd.DataFrame(data=similarity, columns=columns, index=index)

# 결과 확인하기
df_sim
```

```python
### 특정 영화 기준 --> 줄거리가 유사한 영호 추천 함수 생성

def recommend(title):
    top10 = df_sim[title].sort_values(ascending=False)[1:11]
    return top10

# 결과 확인하기
top10 = recommend('GoldenEye')
print(f'결과 확인하기 : \n{top10}')
```
