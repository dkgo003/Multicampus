## 변수

```python
### 변수의 이름 규칙

'''
### 사용 가능한 문자
1. 소문자(a~z)
2. 대문자(A~Z)
3. 숫자(0~9)
4. 한글도 가능 --> 권장 사항(x)
5. 특수기호는 언더바(_) 만 허용

### 규칙
1. 대소 문자를 구분
2. 숫자는 맨 처음에 올 수 없다.
3. 예약어(keyword)는 사용할 수 없다.(for, def, if, import, return, del...)
'''

# 알파벳 (가능) / 대소문자 모두 가능 / 심지어 한글도 가능하나 사용하는 것은 비추!
a = 1
A = 1
변수 = 1
print(a)
print(A)
print(변수)

print('-'*80)

# 알파벳 + 숫자 (가능)
a1 = 2
a2 = 5
print(a1)
print(a2)

print('-'*80)

# 알파벳 + 언더바(_) (가능)
a_ = 1
b_ = 10
print(a_)
print(b_)

print('-'*80)

# 언더바(_) + 알파벳 (가능)
_a = 10
_b = 4
print(_a)
print(_b)

print('-'*80)

# 숫자 + 알파벳 (불가능)
# 1a = 10
# 1_a = 30
# print(1a)
# print(1_a)

# 특수문자 (불가능) --> '-' 사용 가능
# * = 'abc'
# print(*)
_ = 'abc'
print(_)

print('-'*80)

# 변수의 이름 사이의 공백 (불가능)
# a b = 10
# print(a b)
```

### 데이터 타입(자료형)

```python
### 자료형 정리

# 기본형 --> 1개의 데이터
'''
1. int
2. float
3. str
4. bool
5. 자료형 확인 --> 함수 --> type(변수)
'''

# type()을 이용 --> 자료형 확인
print(type(1))
print(type(3.14))
print(type("안녕!"))
print(type("True"))
print(type(True))
```

## 자료형 변환

```python
# int --> float
a = 10
print(f'a의 자료형 = {type(a)}')
b = float(a)
print(f'b의 자료형 = {type(b)}')

print('-'*80)

# float --> int
a = 3.99
print(f'a의 자료형 = {type(a)}')
b = int(a)
print(f'b의 자료형 = {type(b)}')
```

```python
# bool --> int
a = True
print(f'a의 값 = {a}, a의 자료형 = {type(a)}')

print('-'*80)

b = int(a)
print(f'b의 값 = {b}, b의 자료형 = {type(b)}')

print('-'*80)

# bool --> float
a = True
b = float(a)
print(f'b의 값 = {b}, b의 자료형 = {type(b)}')
```
