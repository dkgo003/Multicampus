## 데이터 출력

```python
### 기본 출력 함수 --> print(data)

# 숫자 출력
print(2023)

# 문자열 출력
print('Hello Python!')
print("안녕 파이썬!")
```

```python
### format() 함수 활용 출력

# 숫자 출력
data1 = 180
print("저의 키는 {}cm 입니다.".format(data1))

# 문자열 출력
data2 = '이인규'
print('저의 이름은 {} 입니다.'.format(data2))
```

```python
### f 접두어를 활용한 출력

# 숫자 출력
data1 = 183
print(f'저의 키는 {data1}cm 입니다.')

# 문자열 출력
print(f'저의 이름은 {data2} 입니다.')
```

## 변수와 값

```python
### 변수와 데이터

# 변수 = 데이터(값) ==> 실행 ==> 오른쪽 데이터를 왼쪽 변수에 저장
# 변수의 주소를 알려주는 함수 ==> id(변수)

var = 123
name = 'python'
# data1 = id(var)

print(f"변수 var의 주소 : {id(var)}")
# print(f"변수 var의 주소 : {data1}")
print(f"변수 name의 주소 : {id(name)}")
```

```python
name = "이인규"
print(name)

a = 7
b = 5
print(a + b)

print("a + b")
```

## 기본 자료형

```python
a = 1
b = 1
print(a, b)
print(f"변수 a의 주소 : {id(a)}")

a = 1.5
b = 3.5
print(a, b)
print(f"변수 a의 주소 : {id(a)}")

a = "ABC"
b = "101010"
print(a, b)
print(f"변수 a의 주소 : {id(a)}")

a = True + True
b = False
print(a, b)
print(f"변수 a의 주소 : {id(a)}")
```

## 간단한 연산

```python
### 사직연산

a = 25 + 30
print(a)

b = 30 - 12
print(b)

c = 50 * 3
print(c)

d = 30 / 5
print(d)

print(a * b)
```

```python
### 제곱승(거듭제곱)

a = 3 * 3 * 3 * 3 * 3
print(a)

b = 3 ** 5
print(b)
```

```python
### 나눗셈의 몫과 나머지

# 나누기 2의 몫 추출
a = 7 // 2
print(a)

# 나누기 2의 나머지 추출
b = 7 % 2
print(b)

# 일반적인 나눗셈
c = 7 / 2
print(c)
```

```python
### 증가 연산과 감소 연산

# 증가 연산
a = 1
print(f'처음 변수 a의 주소 : {id(a)}')
a = a + 1
print(a)
print(f'두번째 변수 a의 주소 : {id(a)}')

a += 1
print(a)
print(f'세번째 변수 a의 주소 : {id(a)}')

# 감소 연산
a = a - 1
print(a)
print(f'네번째 변수 a의 주소 : {id(a)}')
a -= 1
print(a)
print(f'다섯번째 변수 a의 주소 : {id(a)}')
```

## 자료형 변환

### 정수형과 실수형 간 변환

```python
### 정수형 --> 실수형

a = 10
print(a)

b = float(10)
print(b)

c = float(a)
print(c)
```

```python
### 실수형 --> 정수형

a = int(10.7)
b = int(10.3)

# 정수형 a와 b의 합을 출력
print(a + b)

# 정수형 a의 값 출력
print(a)

# 정수형 b의 값 출력
print(b)
```

### 숫자형과 문자형 간 변환

```python
### 문자형 --> 숫자형

# 변수 a에 문자 형태의 데이터 76.3 저장 --> '문자형'
a = '76.3'
print(a)

# 문자형을 실수형으로 변환
b = float(a)
print(b)
```

```python
### 숫자형 --> 문자형
a = str(a)
print(a)

b = str(b)
print(b)
print(a + b)
```

### 자료형 확인하기

```python
a = int(10.3)
b = float(10.3)
c = str(10.3)

# 변수 a,b,c의 자료형 확인
print(type(a))
print(type(b))
print(type(c))
```
