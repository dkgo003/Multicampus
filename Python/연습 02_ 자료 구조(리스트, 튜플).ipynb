### 관련 함수

```python
### 추가 --> append() --> 마지막 인덱스에 값을 추가

# 빈 리스트 자료형 생성
mylist = []
print(mylist)

print('-'*80)

# 성분 원소 추가(1)
mylist.append(10)
print(mylist)

print('-'*80)

# 성분 원소 추가(2) --> 중복된 값을 추가할 수 있고, 순서가 유지된다(입력 순서대로 인덱스 생성)
mylist.append(7)
mylist.append(7)
mylist.append(7)
mylist.append(3)
mylist.append(5)
mylist.append(2)
print(mylist)

print('-'*80)

# 성분 원소 추가(3) --> 리스트 자료형 추가 가능
data = [1, 2, 'ab']
mylist.append(data)
print(mylist)
```

```python
### 정렬(sorting) 1

mylist = [1, 6, 3, 2, 7, 5, 4]
print(mylist)

print('-'*80)

# sort() 사용 --> 오름차순으로 정렬 --> 결과 : 원래 변수에 덮어씌우기
# sorted_list = mylist.sort() # 변수로 저장하면 안된다
# print(sorted_list)
mylist.sort()
print(mylist)

print('-'*80)

# 내림차순 정렬 --> sort(reverse=True)
mylist.sort(reverse=True)
print(mylist)
```

```python
### 정렬(sorting) 2

mylist = [1, 6, 3, 2, 7, 5, 4]
print(mylist)

print('-'*80)

# sorted() --> 기본 : 오름차순 정렬, 범용 함수, 결과를 다른 변수에 저장하고 호출
sorted_list = sorted(mylist)
print(sorted_list)

print('-'*80)

# sorted() --> 내림차순 정렬 --> sorted(data, reverse=True)
sorted_list_rev = sorted(sorted_list, reverse=True)
print(sorted_list_rev)
```

```python
### 값 추가 --> 내가 원하는 위치(index)에 값을 추가 --> insert()
mylist = [1, 6, 3, 2, 7, 5, 4]
print(mylist)

print('-'*80)

# index = 3에 [1, 3, '분석가']
data = [1, 3, '분석가']
mylist.insert(3, data)
print(mylist)
```

```python
### 값 추가 --> extend() --> 확장, +연산과 같은 효과
mylist = [1, 2, 3]
print(mylist)

print('-'*80)

# 값 추가 1 --> 1차원 리스트의 형태로 추가
mylist.extend([4, 5, 6])
print(mylist)

print('-'*80)

# 값 추가 2 --> 2차원 리스트의 형태로 추가
mylist.extend([[1, 2, 3]])
print(mylist)
```

```python
### 값 제거 --> remove(data)
mylist = list(range(5))
print(mylist)

print('-'*80)

# remove() 사용
mylist.remove(2)
print(mylist)

print('-'*80)

# remove() 사용
mylist.remove(3)
print(mylist)
```

```python
### 값 제거 --> del --> 특정 index의 값을 제거

mylist = list(range(5))
print(mylist)

print('-'*80)

del mylist[0]
print(mylist)

print('-'*80)

del mylist[0]
print(mylist)
```

```python
### 전체 원소의 갯수 확인 --> len()
mylist = list(range(1, 11 ,2))
print(mylist)

print('-'*80)

# len() 함수 적용
num_elements = len(mylist)
print(num_elements)
```

### 인덱싱

```python
# 리스트 데이터 생성
mylist = ['p', 'y', 't', 'h', 'o', 'n']
print(mylist)

# 인덱싱 --> 특정 원소를 추출
print(f'index=0인 성분 원소 추출 : {mylist[0]}')
print('-'*80)
print(f'마지막 인덱스에 해당 하는 성분 원소 추출 : {mylist[-1]}')
print('-'*80)
print(f'범위를 벗어나는 인덱스 지정하는 경우 : {mylist[100]}')
```

```python
### 인덱스로 접근하여 값 수정하기

mylist = [10, 20, 30, 40 ,50]
print(mylist)

print('-'*80)

# 0번 인덱스에 해당하는 원소의 값을 'abc'로 수정
mylist[0] = 'abc'
print(mylist)

print('-'*80)

# 마지막 인덱스에 해당하는 원소의 값을 999로 수정
mylist[-1] = 999
print(mylist)
```

```python
### 중첩된 리스트(2차원 리스트) 인덱싱

mylist = [['가', '나', '다'], [4, 5, 6], 7, 8 ,9]
print(mylist)

print('-'*80)

# index=1인 성분 원소를 추출
print(mylist[1])

print('-'*80)

# index=1인 성분 원소 --> 두번째 성분 원소를 추출
print(mylist[1][1])
```

### 슬라이싱

```python
### [:] --> 전체 추출
mylist = [10, 20, 30, 40, 50]
print(mylist)

print('-'*80)

# 전체 범위를 지정
print(mylist[:])
```

```python
### [start:] --> start 인덱스부터 끝까지 추출
result = mylist[2:]
print(result)
```

```python
### [:end] --> 처음부터 마지막 성분원소 전까지 추출
print(f'리스트의 성분 원소의 길이 = {len(mylist)}')
result = mylist[:-1]
print(result)
```

```python
### [::2] --> 전체 성분 원소 --> 2칸씩 건너뛰기
result = mylist[::2]
print(result)
```

```python
### 전체 성분 원소에 대한 역순 출력 --> [::-1]
result = mylist[::-1]
print(result)
```

```python
### 전체 성분 원소 --> 역순 & 2칸씩 건너뛰기 --> [::-2]
result = mylist[::-2]
print(result)
```

## 리스트와 튜플간의 변환

```python
### list --> tuple

# 리스트 자료형 데이터 생성
mylist = [1, 2, 3, 4]
print(mylist)

print('-'*80)

# tuple(data)
mytuple = tuple(mylist)
print(f'변환 후 자료형 = {type(mytuple)}')
```

```python
### tuple --> list

# 튜플 자료형 데이터 생성
mytuple = tuple(mylist)
print(mytuple)

print('-'*80)

# list(data)
mylist = list(mytuple)
mylist[0] = 100
print(mylist)
print(f'변환 후 자료형 = {type(mylist)}')
```
