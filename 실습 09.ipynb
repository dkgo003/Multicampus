## 클래스

```python
### [코드 10-2]

class SoccorPlayer:
    def __init__(self, name, position, back_number):
        self.name = name
        self.position = position
        self.back_number = back_number
    
    def change_back_number(self, new_number):
        print(f'선수의 등번호를 변경합니다. : From {self.back_number} to {new_number}')
        self.back_number = new_number

    def __str__(self):
        return f"Hello, My name if {self.name}. I play in {self.position} in center."
        
# def 의 기능 --> 구현

inklee = SoccorPlayer("inklee", "MF", 10)

print(f"현재 선수의 등번호는 : {inklee.back_number}")
inklee.change_back_number(5)
print(f"현재 선수의 등번호는 : {inklee.back_number}")

# 생성된 객체(인스턴스) --> 출력
print(inklee)
```

```python
### [코드 10-3]

# 리스트 자료형 데이터 생성
names = ["Messi", "Ramos", "Ronaldo", "Park", "Buffon"]
positions = ["MF", "DF", "CF", "WF", "GK"]
numbers = [10, 4, 7, 13, 1]

# 이차원 리스트
players = [[name, position, number] for name, position, number in zip(names, positions, numbers)]
print(players)
print(players[0])

print('-'*80)

# 인스턴스 생성
player_instances = [SoccorPlayer(name, position, number) for name, position, number in zip(names, positions, numbers)]
print(player_instances[1])
```

## 상속

```python
### 객체 지향 프로그래밍에서의 상속

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

class Korean(Person):
    pass

first_korean = Korean("Messi", 35)
print(first_korean.name)
```

```python
### [코드 10-4]

class Person:
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender

    def about_me(self):
        print(f'저의 이름은 {self.name}이구요, 제 나이는 {str(self.age)}살 입니다.')

my_id = Person("inklee", 27, "male")
my_id.about_me()
```

```python
### [코드 10-5] 오버리아딩(Overriding)

class Employee(Person):
    def __init__(self, name, age, gender, salary, hire_date):
        super().__init__(name, age, gender)
        self.salary = salary
        self.hire_date = hire_date

    def do_work(self):
        print("열심히 일을 합니다.")

    def about_me(self):
        super().about_me()
        print(f"제 급여는 {self.salary}원 이구요, 제 입사일은 {self.hire_date}입니다.")

# 객체 생성
myEmployee = Employee('inklee', '27', 'name', 5000000, "2023/3/21")

# 객체 실행
myEmployee.about_me()
```

```python
### [코드 10-7] 오버라이딩 + 추상화

class Animal:
    def __init__(self, name):
        self.name = name
        
    def talk(self):
        raise NotImplementedError("Subclass must implement abstractmethod")

class Cat(Animal):
    def talk(self):
        return 'Meow!'

class Dog(Animal):
    def talk(self):
        return 'Woof! Woof!'

animals = [Cat('Missy'), Cat('Mr. Mistoffelees'), Dog('Lassie')]

for animal in animals:
    print(animal.name + ': ' + animal.talk())
```
